## **🧾 G. Cấu trúc Rẽ nhánh trong Python (`if`, `elif`, `else`)**

Trong lập trình, đôi khi ta cần **đưa ra quyết định** – ví dụ: nếu người dùng đủ tuổi thì cho truy cập, nếu không thì từ chối.

🔹 Đó là lúc ta cần đến **cấu trúc rẽ nhánh** – giúp chương trình **chạy theo nhiều hướng khác nhau** dựa trên điều kiện cụ thể.

Python hỗ trợ rẽ nhánh thông qua các **từ khóa đặc biệt**: `if`, `elif`, `else`.

| **Từ khóa / Cú pháp**     | **Chức năng**                                                     | **Ví dụ sử dụng thực tế**                                 |
|---------------------------|--------------------------------------------------------------------|-------------------------------------------------------------|
| `if điều_kiện:`           | Kiểm tra điều kiện đơn                                            | `if score >= 5:`                                           |
| `if ... elif ... else:`   | Rẽ nhánh nhiều trường hợp                                         | Kiểm tra điểm và phân loại xếp loại                        |
| `if biểu_thức:`           | Cho phép dùng mọi kiểu dữ liệu – Python tự đánh giá True/False   | `if "abc":`, `if []:`                                     |
| `A if cond else B`        | Cú pháp rút gọn (inline if)                                       | `msg = "Đậu" if score >= 5 else "Rớt"`                     |


📌 Những từ khóa này là **keyword** của Python, **không phải hàm**, cũng không phải toán tử.  
📌 Việc **thụt lề (indent)** là **bắt buộc** để xác định khối lệnh đi kèm.

📌 Cú pháp tổng quát:
```python
if điều_kiện:
    # khối lệnh nếu đúng
elif điều_kiện_khác:
    # khối lệnh nếu điều_kiện_khác đúng
else:
    # khối lệnh nếu tất cả điều kiện đều sai



#### 📦 **Đặc điểm nổi bật của `if`, `elif`, `else`**

Những đặc điểm quan trọng giúp cấu trúc rẽ nhánh:

1. ✅ **Kiểm tra điều kiện để ra quyết định**  
2. ✅ **Có thể mở rộng với nhiều `elif` cho các trường hợp khác nhau**  
3. ✅ **`else` là phần mặc định khi không điều kiện nào đúng**  
4. ✅ **Cú pháp dễ đọc, sử dụng `:` và thụt lề để xác định khối lệnh**  
5. ✅ **Điều kiện trong `if`, `elif` có thể là bất kỳ biểu thức nào có kết quả tương đương Boolean**  
6. ⚠️ **Không bắt buộc là `True`/`False` rõ ràng – Python tự đánh giá theo “truthy/falsy”**

---

📌 Cụ thể:

- Các kiểu dữ liệu như `0`, `None`, `''`, `[]`, `{}` được coi là **False**  
- Các giá trị như `1`, `"abc"`, `[1, 2]`, `{1:2}` được coi là **True**


#### 📦 **Ví Dụ**

````python
if 1:       # được đánh giá là True
    print("Sẽ chạy")

if 0:       # được đánh giá là False
    print("Không chạy")
````

````python
# ✅ 1. if (điều_kiện)
score = 6

# Nếu điểm lớn hơn hoặc bằng 5 thì in ra "daiduong.eth"
if score >= 5:
    print("daiduong.eth")
````

````python
# ✅ 2. if ... elif ... else
score = 7

# Kiểm tra điểm và phân loại xếp loại
if score >= 8:
    print("Giỏi")
elif score >= 5:
    print("Trung bình")
else:
    print("Kém")
````

````python
# ✅ 3. if biểu_thức (truthy / falsy)
name = ""

# Nếu name là chuỗi rỗng thì Python sẽ hiểu là False
if name:
    print("Tên đã được nhập")
else:
    print("Chưa nhập tên")
````

````python
# ✅ 4. Cú pháp rút gọn: A if điều_kiện else B
score = 4

# Gán "Đậu" nếu score >= 5, ngược lại là "Rớt"
result = "Đậu" if score >= 5 else "Rớt"
print(result)
````

````python
# ✅ 5. Lồng if trong if
x = 7

# Kiểm tra đồng thời 2 điều kiện: > 0 và < 10
if x > 0:
    if x < 10:
        print("x là số dương nhỏ hơn 10")
````

````python
# ✅ 6. Kết hợp nhiều điều kiện với and
age = 20
has_id = True

# Chỉ cho vào nếu đủ tuổi và có giấy tờ
if age >= 18 and has_id:
    print("Đủ điều kiện vào cổng")
````

## **🧾 H. Cấu trúc Lặp trong Python (`for`, `while`)**

Trong lập trình, có những lúc bạn cần **thực hiện một hành động nhiều lần** – ví dụ: in ra từng phần tử trong List, xử lý từng dòng dữ liệu,...

🔹 Đó là lúc ta cần đến **vòng lặp** – giúp chương trình **tự động lặp lại thao tác** cho đến khi điều kiện kết thúc.

Python hỗ trợ hai loại vòng lặp cơ bản:

| **Từ khóa / Cú pháp** | **Chức năng**                                                       | **Ví dụ sử dụng thực tế**                          |
|------------------------|----------------------------------------------------------------------|----------------------------------------------------|
| `for ... in ...:`      | Lặp qua từng phần tử trong một chuỗi, list, tuple, set, dict,...     | Duyệt danh sách học sinh, in từng ký tự của chuỗi |
| `while điều_kiện:`     | Lặp lại cho đến khi điều kiện không còn đúng nữa                     | Đếm số lần nhập mật khẩu cho đến khi đúng         |

📌 Cả `for` và `while` đều là **từ khóa của Python**, không phải hàm.  
📌 Phần thân vòng lặp **phải thụt lề** đúng theo quy định.  
📌 Có thể kết hợp với từ khóa `break`, `continue` để kiểm soát vòng lặp.

📌 Cú pháp tổng quát:

```python
# for loop
for biến in tập_hợp:
    # khối lệnh

# while loop
while điều_kiện:
    # khối lệnh


#### 📦 **Đặc điểm nổi bật của `for` và `while`**

Những đặc điểm quan trọng giúp vòng lặp trở thành công cụ quan trọng trong lập trình:

1. ✅ **Tự động lặp lại khối lệnh nhiều lần**  
2. ✅ **`for` dùng để lặp qua từng phần tử trong tập hợp** (`list`, `tuple`, `str`, `set`, `dict`, ...)  
3. ✅ **`while` lặp khi điều kiện còn đúng** – dùng khi **chưa biết trước số lần lặp**  
4. ✅ **Có thể dùng `break` để thoát vòng lặp sớm**  
5. ✅ **Có thể dùng `continue` để bỏ qua vòng lặp hiện tại và sang vòng tiếp theo**  
6. ⚠️ **Cần tránh lặp vô hạn (infinite loop)** nếu dùng `while` mà không cập nhật điều kiện

---

📌 So sánh nhanh:

| Đặc điểm             | `for` loop                       | `while` loop                          |
|----------------------|----------------------------------|----------------------------------------|
| Lặp qua tập hợp      | ✅                               | ❌                                    |
| Kiểm tra điều kiện   | Ẩn bên trong                      | Do người lập trình tự kiểm soát       |
| Biết trước số lần    | Thường có                        | Không chắc – tùy điều kiện             |
| Dễ gây lặp vô hạn    | Rất hiếm                         | Có thể xảy ra nếu không cẩn thận       |


#### 📚 **Các Cú pháp/Từ khóa phổ biến với `for` và `while`**

| **Cú pháp/Từ khóa**                     | **Ý nghĩa / Chức năng**                                               | **Ví dụ sử dụng thực tế**                             |
|---------------------------------|------------------------------------------------------------------------|--------------------------------------------------------|
| `for x in iterable:`           | Duyệt từng phần tử trong một chuỗi / list / set / tuple / dict key    | `for x in [1,2,3]:` → In từng số                      |
| `for k, v in dict.items():`    | Duyệt cả key và value trong dictionary                                | `for k, v in my_dict.items():`                        |
| `for i in range(n):`           | Lặp `n` lần với chỉ số từ 0 đến `n-1`                                  | `for i in range(5):` → Lặp 5 lần                      |
| `while điều_kiện:`             | Lặp đến khi điều kiện sai                                              | `while x < 10:`                                       |
| `break`                        | Thoát khỏi vòng lặp ngay lập tức                                       | `if x == 5: break`                                    |
| `continue`                     | Bỏ qua phần còn lại và sang vòng tiếp theo                             | `if x % 2 == 0: continue`                             |

📌 Lưu ý:
- `for` hoạt động tốt với các **iterable** như: `list`, `tuple`, `set`, `str`, `dict`
- `while` dùng khi bạn **không biết trước số lần lặp**
- Cả hai đều yêu cầu **dấu `:` và thụt lề (indent)** đúng chuẩn Python


#### 📦 **Ví Dụ**

````python
# ✅ 1. for x in iterable
fruits = ["apple", "banana", "cherry"]

# Lặp qua từng phần tử trong danh sách
for fruit in fruits:
    print(fruit)
````

````python
# ✅ 2. for k, v in dict.items()
info = {"name": "Ngọc", "age": 24}

# Duyệt cả key và value trong dict
for key, value in info.items():
    print(key, "→", value)
````

````python
# ✅ 3. for i in range(n)
# Lặp 5 lần từ 0 đến 4
for i in range(5):
    print("Lần thứ", i)
````

````python
# ✅ 4. while điều_kiện
x = 0

# Lặp cho đến khi x >= 3
while x < 3:
    print("x =", x)
    x += 1
````

````python
# ✅ 5. break – thoát vòng lặp
for num in range(10):
    if num == 5:
        break  # Dừng vòng lặp khi gặp số 5
    print(num)
````

````python
# ✅ 6. continue – bỏ qua vòng hiện tại
for num in range(5):
    if num == 2:
        continue  # Bỏ qua 2
    print(num)
````
