````python
# âœ… 1. if (Ä‘iá»u_kiá»‡n)
score = 6

# Náº¿u Ä‘iá»ƒm lá»›n hÆ¡n hoáº·c báº±ng 5 thÃ¬ in ra "daiduong.eth"
if score >= 5:
    print("daiduong.eth")
````

````python
# âœ… 2. if ... elif ... else
score = 7

# Kiá»ƒm tra Ä‘iá»ƒm vÃ  phÃ¢n loáº¡i xáº¿p loáº¡i
if score >= 8:
    print("Giá»i")
elif score >= 5:
    print("Trung bÃ¬nh")
else:
    print("KÃ©m")
````

````python
# âœ… 3. if biá»ƒu_thá»©c (truthy / falsy)
name = ""

# Náº¿u name lÃ  chuá»—i rá»—ng thÃ¬ Python sáº½ hiá»ƒu lÃ  False
if name:
    print("TÃªn Ä‘Ã£ Ä‘Æ°á»£c nháº­p")
else:
    print("ChÆ°a nháº­p tÃªn")
````

````python
# âœ… 4. CÃº phÃ¡p rÃºt gá»n: A if Ä‘iá»u_kiá»‡n else B
score = 4

# GÃ¡n "Äáº­u" náº¿u score >= 5, ngÆ°á»£c láº¡i lÃ  "Rá»›t"
result = "Äáº­u" if score >= 5 else "Rá»›t"
print(result)
````

````python
# âœ… 5. Lá»“ng if trong if
x = 7

# Kiá»ƒm tra Ä‘á»“ng thá»i 2 Ä‘iá»u kiá»‡n: > 0 vÃ  < 10
if x > 0:
    if x < 10:
        print("x lÃ  sá»‘ dÆ°Æ¡ng nhá» hÆ¡n 10")
````

````python
# âœ… 6. Káº¿t há»£p nhiá»u Ä‘iá»u kiá»‡n vá»›i and
age = 20
has_id = True

# Chá»‰ cho vÃ o náº¿u Ä‘á»§ tuá»•i vÃ  cÃ³ giáº¥y tá»
if age >= 18 and has_id:
    print("Äá»§ Ä‘iá»u kiá»‡n vÃ o cá»•ng")
````

| **Tá»« khÃ³a / CÃº phÃ¡p** | **Chá»©c nÄƒng**                                                       | **VÃ­ dá»¥ sá»­ dá»¥ng thá»±c táº¿**                          |

#### ğŸ“¦ **Äáº·c Ä‘iá»ƒm ná»•i báº­t cá»§a `for` vÃ  `while`**

Nhá»¯ng Ä‘áº·c Ä‘iá»ƒm quan trá»ng giÃºp vÃ²ng láº·p trá»Ÿ thÃ nh cÃ´ng cá»¥ quan trá»ng trong láº­p trÃ¬nh:

1. âœ… **Tá»± Ä‘á»™ng láº·p láº¡i khá»‘i lá»‡nh nhiá»u láº§n**  
2. âœ… **`for` dÃ¹ng Ä‘á»ƒ láº·p qua tá»«ng pháº§n tá»­ trong táº­p há»£p** (`list`, `tuple`, `str`, `set`, `dict`, ...)  
3. âœ… **`while` láº·p khi Ä‘iá»u kiá»‡n cÃ²n Ä‘Ãºng** â€“ dÃ¹ng khi **chÆ°a biáº¿t trÆ°á»›c sá»‘ láº§n láº·p**  
4. âœ… **CÃ³ thá»ƒ dÃ¹ng `break` Ä‘á»ƒ thoÃ¡t vÃ²ng láº·p sá»›m**  
5. âœ… **CÃ³ thá»ƒ dÃ¹ng `continue` Ä‘á»ƒ bá» qua vÃ²ng láº·p hiá»‡n táº¡i vÃ  sang vÃ²ng tiáº¿p theo**  
6. âš ï¸ **Cáº§n trÃ¡nh láº·p vÃ´ háº¡n (infinite loop)** náº¿u dÃ¹ng `while` mÃ  khÃ´ng cáº­p nháº­t Ä‘iá»u kiá»‡n

---

ğŸ“Œ So sÃ¡nh nhanh:

| Äáº·c Ä‘iá»ƒm             | `for` loop                       | `while` loop                          |
|----------------------|----------------------------------|----------------------------------------|
| Láº·p qua táº­p há»£p      | âœ…                               | âŒ                                    |
| Kiá»ƒm tra Ä‘iá»u kiá»‡n   | áº¨n bÃªn trong                      | Do ngÆ°á»i láº­p trÃ¬nh tá»± kiá»ƒm soÃ¡t       |
| Biáº¿t trÆ°á»›c sá»‘ láº§n    | ThÆ°á»ng cÃ³                        | KhÃ´ng cháº¯c â€“ tÃ¹y Ä‘iá»u kiá»‡n             |
| Dá»… gÃ¢y láº·p vÃ´ háº¡n    | Ráº¥t hiáº¿m                         | CÃ³ thá»ƒ xáº£y ra náº¿u khÃ´ng cáº©n tháº­n       |


#### ğŸ“š **CÃ¡c CÃº phÃ¡p/Tá»« khÃ³a phá»• biáº¿n vá»›i `for` vÃ  `while`**

| **CÃº phÃ¡p/Tá»« khÃ³a**                     | **Ã nghÄ©a / Chá»©c nÄƒng**                                               | **VÃ­ dá»¥ sá»­ dá»¥ng thá»±c táº¿**                             |
|---------------------------------|------------------------------------------------------------------------|--------------------------------------------------------|
| `for x in iterable:`           | Duyá»‡t tá»«ng pháº§n tá»­ trong má»™t chuá»—i / list / set / tuple / dict key    | `for x in [1,2,3]:` â†’ In tá»«ng sá»‘                      |
| `for k, v in dict.items():`    | Duyá»‡t cáº£ key vÃ  value trong dictionary                                | `for k, v in my_dict.items():`                        |
| `for i in range(n):`           | Láº·p `n` láº§n vá»›i chá»‰ sá»‘ tá»« 0 Ä‘áº¿n `n-1`                                  | `for i in range(5):` â†’ Láº·p 5 láº§n                      |
| `while Ä‘iá»u_kiá»‡n:`             | Láº·p Ä‘áº¿n khi Ä‘iá»u kiá»‡n sai                                              | `while x < 10:`                                       |
| `break`                        | ThoÃ¡t khá»i vÃ²ng láº·p ngay láº­p tá»©c                                       | `if x == 5: break`                                    |
| `continue`                     | Bá» qua pháº§n cÃ²n láº¡i vÃ  sang vÃ²ng tiáº¿p theo                             | `if x % 2 == 0: continue`                             |

ğŸ“Œ LÆ°u Ã½:
- `for` hoáº¡t Ä‘á»™ng tá»‘t vá»›i cÃ¡c **iterable** nhÆ°: `list`, `tuple`, `set`, `str`, `dict`
- `while` dÃ¹ng khi báº¡n **khÃ´ng biáº¿t trÆ°á»›c sá»‘ láº§n láº·p**
- Cáº£ hai Ä‘á»u yÃªu cáº§u **dáº¥u `:` vÃ  thá»¥t lá» (indent)** Ä‘Ãºng chuáº©n Python


#### ğŸ“¦ **VÃ­ Dá»¥**

````python
# âœ… 1. for x in iterable
fruits = ["apple", "banana", "cherry"]

# Láº·p qua tá»«ng pháº§n tá»­ trong danh sÃ¡ch
for fruit in fruits:
    print(fruit)
````

````python
# âœ… 2. for k, v in dict.items()
info = {"name": "Ngá»c", "age": 24}

# Duyá»‡t cáº£ key vÃ  value trong dict
for key, value in info.items():
    print(key, "â†’", value)
````

````python
# âœ… 3. for i in range(n)
# Láº·p 5 láº§n tá»« 0 Ä‘áº¿n 4
for i in range(5):
    print("Láº§n thá»©", i)
````

````python
# âœ… 4. while Ä‘iá»u_kiá»‡n
x = 0

# Láº·p cho Ä‘áº¿n khi x >= 3
while x < 3:
    print("x =", x)
    x += 1
````

````python
# âœ… 5. break â€“ thoÃ¡t vÃ²ng láº·p
for num in range(10):
    if num == 5:
        break  # Dá»«ng vÃ²ng láº·p khi gáº·p sá»‘ 5
    print(num)
````

````python
# âœ… 6. continue â€“ bá» qua vÃ²ng hiá»‡n táº¡i
for num in range(5):
    if num == 2:
        continue  # Bá» qua 2
    print(num)
````
