## **üßæ E_01. Python Operators (To√°n T·ª≠ trong Python)**

**To√°n t·ª≠ (operators)** l√† c√°c k√Ω hi·ªáu d√πng ƒë·ªÉ th·ª±c hi·ªán c√°c ph√©p t√≠nh ho·∫∑c so s√°nh gi·ªØa c√°c gi√° tr·ªã.

Trong Python, ch√∫ng ta th∆∞·ªùng d√πng c√°c lo·∫°i to√°n t·ª≠ sau:
- To√°n t·ª≠ s·ªë h·ªçc
- To√°n t·ª≠ so s√°nh
- To√°n t·ª≠ logic
- To√°n t·ª≠ g√°n

#### ‚úÖ 1. To√°n T·ª≠ S·ªë H·ªçc

D√πng ƒë·ªÉ th·ª±c hi·ªán c√°c ph√©p t√≠nh c∆° b·∫£n nh∆∞ c·ªông, tr·ª´, nh√¢n, chia...

| To√°n t·ª≠ | √ù nghƒ©a        | V√≠ d·ª•        |
|--------|----------------|--------------|
| `+`    | C·ªông           | `a + b`      |
| `-`    | Tr·ª´            | `a - b`      |
| `*`    | Nh√¢n           | `a * b`      |
| `/`    | Chia           | `a / b`      |
| `//`   | Chia l·∫•y nguy√™n | `a // b`     |
| `%`    | Chia l·∫•y d∆∞    | `a % b`      |
| `**`   | L≈©y th·ª´a       | `a ** b`     |


````python
a = 7
b = 3

print("C·ªông:", a + b)
print("Tr·ª´:", a - b)
print("Nh√¢n:", a * b)
print("Chia:", a / b)
print("Chia l·∫•y ph·∫ßn nguy√™n:", a // b)
print("Chia l·∫•y d∆∞:", a % b)
print("L≈©y th·ª´a:", a ** b)
````

#### ‚úÖ 2. To√°n T·ª≠ So S√°nh
D√πng ƒë·ªÉ so s√°nh hai gi√° tr·ªã, k·∫øt qu·∫£ tr·∫£ v·ªÅ l√† `True` ho·∫∑c `False`.

| To√°n t·ª≠ | √ù nghƒ©a              |
|--------|-----------------------|
| `==`   | b·∫±ng nhau             |
| `!=`   | kh√¥ng b·∫±ng nhau       |
| `>`    | l·ªõn h∆°n               |
| `<`    | nh·ªè h∆°n               |
| `>=`   | l·ªõn h∆°n ho·∫∑c b·∫±ng     |
| `<=`   | nh·ªè h∆°n ho·∫∑c b·∫±ng     |

````python
x = 22
y = 10

print(x == y)   # b·∫±ng
print(x != y)   # kh√¥ng b·∫±ng
print(x > y)    # l·ªõn h∆°n
print(x < y)    # nh·ªè h∆°n
print(x >= y)   # l·ªõn h∆°n ho·∫∑c b·∫±ng
print(x <= y)   # nh·ªè h∆°n ho·∫∑c b·∫±ng
````

#### ‚úÖ 3. To√°n T·ª≠ Logic
D√πng ƒë·ªÉ k·∫øt h·ª£p c√°c ƒëi·ªÅu ki·ªán (boolean).

| To√°n t·ª≠ | √ù nghƒ©a                     |
|--------|------------------------------|
| `and`  | c·∫£ hai ƒëi·ªÅu ki·ªán ƒë·ªÅu ƒë√∫ng   |
| `or`   | m·ªôt trong hai ƒë√∫ng l√† ƒë∆∞·ª£c |
| `not`  | ph·ªß ƒë·ªãnh ƒëi·ªÅu ki·ªán          |

````python
a = True
b = False

print(a and b)  # c·∫£ hai ph·∫£i ƒë√∫ng
print(a or b)   # m·ªôt trong hai ƒë√∫ng l√† ƒë∆∞·ª£c
print(not a)    # ph·ªß ƒë·ªãnh
````

#### ‚úÖ 4. To√°n T·ª≠ G√°n
D√πng ƒë·ªÉ g√°n gi√° tr·ªã cho bi·∫øn. C√≥ th·ªÉ k·∫øt h·ª£p v·ªõi c√°c to√°n t·ª≠ kh√°c.

| To√°n t·ª≠ | V√≠ d·ª•      | √ù nghƒ©a                  |
|--------|------------|---------------------------|
| `=`    | `x = 5`    | g√°n gi√° tr·ªã 5 cho x       |
| `+=`   | `x += 1`   | x = x + 1                  |
| `-=`   | `x -= 1`   | x = x - 1                  |
| `*=`   | `x *= 2`   | x = x * 2                  |
| `/=`   | `x /= 2`   | x = x / 2                  |

````python
# G√°n gi√° tr·ªã ban ƒë·∫ßu
x = 123
print("Ban ƒë·∫ßu x =", x)

# += : c·ªông r·ªìi g√°n
x += 5    # t∆∞∆°ng ƒë∆∞∆°ng x = x + 5
print("Sau x += 5:", x)

# -= : tr·ª´ r·ªìi g√°n
x -= 3    # t∆∞∆°ng ƒë∆∞∆°ng x = x - 3
print("Sau x -= 3:", x)

# *= : nh√¢n r·ªìi g√°n
x *= 2    # t∆∞∆°ng ƒë∆∞∆°ng x = x * 2
print("Sau x *= 2:", x)

# /= : chia r·ªìi g√°n
x /= 4    # t∆∞∆°ng ƒë∆∞∆°ng x = x / 4
print("Sau x /= 4:", x)

# //= : chia l·∫•y nguy√™n r·ªìi g√°n
x //= 2   # t∆∞∆°ng ƒë∆∞∆°ng x = x // 2
print("Sau x //= 2:", x)

# %= : chia l·∫•y d∆∞ r·ªìi g√°n
x %= 4    # t∆∞∆°ng ƒë∆∞∆°ng x = x % 4
print("Sau x %= 3:", x)

# **= : l≈©y th·ª´a r·ªìi g√°n
x **= 2   # t∆∞∆°ng ƒë∆∞∆°ng x = x ** 2
print("Sau x **= 2:", x)
````
