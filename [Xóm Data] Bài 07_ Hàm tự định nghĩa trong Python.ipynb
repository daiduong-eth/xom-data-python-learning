## **üìö F_02. User-Defined Function (H√†m t·ª± ƒë·ªãnh nghƒ©a trong Python)**

Trong Python, **h√†m (function)** l√† c√¥ng c·ª• gi√∫p b·∫°n ƒë√≥ng g√≥i m·ªôt ƒëo·∫°n m√£ th·ª±c hi·ªán m·ªôt c√¥ng vi·ªác c·ª• th·ªÉ. Sau ƒë√≥ c√≥ th·ªÉ **tr·∫£ v·ªÅ m·ªôt gi√° tr·ªã n√†o ƒë√≥ (return value)**.

üëâ Trong ph·∫ßn F_02 n√†y, ch√∫ng ta s·∫Ω h·ªçc c√°ch **t·ª± ƒë·ªãnh nghƒ©a (define) c√°c h√†m c·ªßa ch√≠nh m√¨nh** ‚Äî m·ªôt k·ªπ nƒÉng quan tr·ªçng khi l·∫≠p tr√¨nh th·ª±c t·∫ø.




## **üìå C√∫ ph√°p c∆° b·∫£n c·ªßa h√†m t·ª± ƒë·ªãnh nghƒ©a**

ƒê·ªÉ ƒë·ªãnh nghƒ©a m·ªôt h√†m trong Python, ch√∫ng ta s·ª≠ d·ª•ng t·ª´ kh√≥a `def`:

```python
def ten_ham(tham_so_1, tham_so_2, ...):
    # Kh·ªëi l·ªánh x·ª≠ l√Ω b√™n trong h√†m
    # (c√≥ th·ªÉ c√≥ ho·∫∑c kh√¥ng)
    return gia_tri_tra_ve
```

Gi·∫£i th√≠ch:

`def` l√† t·ª´ kh√≥a b·∫Øt bu·ªôc khi ƒë·ªãnh nghƒ©a h√†m.

`ten_ham` l√† t√™n do b·∫°n t·ª± ƒë·∫∑t (theo quy t·∫Øc ƒë·∫∑t t√™n bi·∫øn).

Trong ngo·∫∑c `()` l√† danh s√°ch c√°c tham s·ªë (c√≥ th·ªÉ c√≥ ho·∫∑c kh√¥ng).

`return` tr·∫£ v·ªÅ gi√° tr·ªã ra b√™n ngo√†i h√†m.


````python
# ƒê·ªãnh nghƒ©a h√†m c·ªông hai s·ªë
def cong_hai_so(a, b):
    tong = a + b
    return tong

# G·ªçi h√†m
ket_qua = cong_hai_so(3, 5)
print(ket_qua)
````

#### üì¶ **M·ªôt v√†i ƒë·∫∑c ƒëi·ªÉm c·ªßa h√†m t·ª± ƒë·ªãnh nghƒ©a trong Python**

1. ‚úÖ H√†m **c√≥ th·ªÉ c√≥ ho·∫∑c kh√¥ng c√≥ tham s·ªë truy·ªÅn v√†o**.
2. ‚úÖ H√†m **c√≥ th·ªÉ c√≥ ho·∫∑c kh√¥ng c√≥ gi√° tr·ªã tr·∫£ v·ªÅ (`return`)**.
3. ‚úÖ Bi·∫øn b√™n trong h√†m c√≥ ph·∫°m vi ri√™ng (**local scope**), ch·ªâ t·ªìn t·∫°i trong h√†m.


#### **V√≠ D·ª•**

````python
### 1Ô∏è‚É£ V√≠ d·ª• h√†m **kh√¥ng c√≥ tham s·ªë**

def xin_chao():
    print("Xin ch√†o c√°c b·∫°n t·ª´ X√≥m Data!")

# G·ªçi h√†m
xin_chao()

###‚úÖ Gi·∫£i th√≠ch:
###    - H√†m xin_chao() kh√¥ng nh·∫≠n b·∫•t k·ª≥ tham s·ªë n√†o.
###    - Khi g·ªçi h√†m, ch·ªâ c·∫ßn ghi xin_chao() l√† ƒë·ªß.
````

````python
### 2Ô∏è‚É£ V√≠ d·ª• h√†m **c√≥ tham s·ªë**

def tinh_tong(a, b):
    return a + b

# G·ªçi h√†m v·ªõi ƒë·ªëi s·ªë
ket_qua = tinh_tong(3, 5)
print("T·ªïng c·ªßa 3 v√† 5 l√†:", ket_qua)

###‚úÖ Gi·∫£i th√≠ch:
###  - H√†m tinh_tong() nh·∫≠n 2 tham s·ªë a v√† b.
###  - Khi g·ªçi h√†m, c·∫ßn truy·ªÅn v√†o ƒë·ªß s·ªë l∆∞·ª£ng ƒë·ªëi s·ªë t∆∞∆°ng ·ª©ng.
````

````python
### 3Ô∏è‚É£ V√≠ d·ª• h√†m **kh√¥ng c√≥ gi√° tr·ªã tr·∫£ v·ªÅ (`return`)**

def thong_bao():
    print("H√†m n√†y ch·ªâ in ra th√¥ng b√°o, kh√¥ng c√≥ return.")

# G·ªçi h√†m
thong_bao()

### ‚úÖ Gi·∫£i th√≠ch:
### - H√†m th·ª±c hi·ªán h√†nh ƒë·ªông in th√¥ng b√°o.
### - Kh√¥ng c√≥ c√¢u l·ªánh `return`, n√™n h√†m t·ª± ƒë·ªông tr·∫£ v·ªÅ `None`.
````

````python
### 4Ô∏è‚É£ V√≠ d·ª• h√†m **c√≥ gi√° tr·ªã tr·∫£ v·ªÅ (`return`)**

def tinh_hieu(a, b):
    return a - b

# G·ªçi h√†m v√† nh·∫≠n gi√° tr·ªã tr·∫£ v·ªÅ
ket_qua = tinh_hieu(10, 3)
print("Hi·ªáu c·ªßa 10 v√† 3 l√†:", ket_qua)

### ‚úÖ Gi·∫£i th√≠ch:
### - H√†m c√≥ c√¢u l·ªánh `return` ƒë·ªÉ tr·∫£ v·ªÅ k·∫øt qu·∫£.
### - C√≥ th·ªÉ g√°n k·∫øt qu·∫£ tr·∫£ v·ªÅ cho bi·∫øn v√† s·ª≠ d·ª•ng ti·∫øp.
````

````python
### 5Ô∏è‚É£ V√≠ d·ª• h√†m **kh√¥ng c√≥ tham s·ªë nh∆∞ng c√≥ gi√° tr·ªã tr·∫£ v·ªÅ (`return`)**

def get_xom_data_name():
    return "X√≥m Data"

# G·ªçi h√†m v√† nh·∫≠n gi√° tr·ªã tr·∫£ v·ªÅ
ten_group = get_xom_data_name()
print("T√™n nh√≥m l√†:", ten_group)

### ‚úÖ Gi·∫£i th√≠ch:
### - H√†m kh√¥ng c·∫ßn tham s·ªë ƒë·∫ßu v√†o.
### - H√†m v·∫´n tr·∫£ v·ªÅ gi√° tr·ªã b·∫±ng c√¢u l·ªánh `return`.
````

````python
### 6Ô∏è‚É£ V√≠ d·ª• v·ªÅ **ph·∫°m vi bi·∫øn trong h√†m (Local Scope)**

def demo_scope():
    x = 10
    print("Gi√° tr·ªã x trong h√†m:", x)

# G·ªçi h√†m
demo_scope()

print(x)  # L·ªói: NameError v√¨ x kh√¥ng t·ªìn t·∫°i b√™n ngo√†i h√†m

### ‚úÖ Gi·∫£i th√≠ch:
### - Bi·∫øn `x` ch·ªâ t·ªìn t·∫°i trong ph·∫°m vi c·ªßa h√†m (`local scope`).
### - B√™n ngo√†i h√†m kh√¥ng th·ªÉ truy c·∫≠p ƒë∆∞·ª£c bi·∫øn `x`.
````

````python
### 7Ô∏è‚É£ V√≠ d·ª• v·ªÅ **bi·∫øn to√†n c·ª•c (Global Scope)**

x = 22  # Bi·∫øn to√†n c·ª•c

def demo_global_1():
    x = 10
    print("Gi√° tr·ªã x to√†n c·ª•c b√™n trong h√†m:", x)

def demo_global_2():
    print("Gi√° tr·ªã x to√†n c·ª•c b√™n trong h√†m:", x)

# G·ªçi h√†m
demo_global_1()
demo_global_2()
print("Gi√° tr·ªã x to√†n c·ª•c b√™n ngo√†i h√†m:", x)

### ‚úÖ Gi·∫£i th√≠ch:
### - Bi·∫øn `x` ƒë∆∞·ª£c khai b√°o b√™n ngo√†i t·∫•t c·∫£ c√°c h√†m ‚Üí global scope.
### - C√°c h√†m c√≥ th·ªÉ ƒë·ªçc ƒë∆∞·ª£c gi√° tr·ªã global n·∫øu kh√¥ng ghi ƒë√®.
````
