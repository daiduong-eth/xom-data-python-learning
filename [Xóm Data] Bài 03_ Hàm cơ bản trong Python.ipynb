## **üìö F_01. Python Function (H√†m trong Python)**

Trong Python, **h√†m (function)** l√† m·ªôt c√¥ng c·ª• gi√∫p b·∫°n **th·ª±c hi·ªán m·ªôt c√¥ng vi·ªác c·ª• th·ªÉ**: x·ª≠ l√Ω d·ªØ li·ªáu, t√≠nh to√°n, chuy·ªÉn ƒë·ªïi ki·ªÉu, ki·ªÉm tra ƒëi·ªÅu ki·ªán...
sau ƒë√≥ **tr·∫£ v·ªÅ m·ªôt gi√° tr·ªã n√†o ƒë√≥**.

#### ‚úÖ 1. ƒê·∫∑c ƒëi·ªÉm c·ªßa h√†m

1. **Th·ª±c hi·ªán m·ªôt h√†nh ƒë·ªông c·ª• th·ªÉ**  
   ‚Üí V√≠ d·ª•: `print()` ƒë·ªÉ in ra m√†n h√¨nh, `type()` ƒë·ªÉ ki·ªÉm tra ki·ªÉu, `abs()` ƒë·ªÉ l·∫•y gi√° tr·ªã tuy·ªát ƒë·ªëi c·ªßa c√°c gi√° tr·ªã s·ªë.

2. **C√≥ th·ªÉ nh·∫≠n ƒë·∫ßu v√†o (g·ªçi l√† ƒë·ªëi s·ªë ‚Äì arguments)**  
   ‚Üí Khi b·∫°n s·ª≠ d·ª•ng m·ªôt h√†m, th∆∞·ªùng b·∫°n s·∫Ω **truy·ªÅn d·ªØ li·ªáu v√†o trong d·∫•u ngo·∫∑c** c·ªßa h√†m ƒë√≥. D·ªØ li·ªáu ƒë∆∞·ª£c truy·ªÅn v√†o n√†y ƒë∆∞·ª£c g·ªçi l√† **ƒë·ªëi s·ªë** (argument).

   V√≠ d·ª•: `len("Hello")`  
   - `len` l√† t√™n c·ªßa **h√†m** ‚Äì c√≥ nhi·ªám v·ª• ƒëo ƒë·ªô d√†i c·ªßa m·ªôt ƒë·ªëi t∆∞·ª£ng (chu·ªói, danh s√°ch, v.v.)
   - `"Hello"` l√† m·ªôt **gi√° tr·ªã c√≥ ki·ªÉu (string)** ƒë∆∞·ª£c truy·ªÅn v√†o h√†m ‚Äì ch√≠nh l√† **ƒë·ªëi s·ªë**  
   ‚Üí Khi ch·∫°y `len("Hello")`, Python s·∫Ω tr·∫£ v·ªÅ k·∫øt qu·∫£ **5**, v√¨ chu·ªói `"Hello"` c√≥ 5 k√Ω t·ª±.

   üß† Kh√¥ng ph·∫£i h√†m n√†o c≈©ng c·∫ßn ƒë·ªëi s·ªë:
   - C√≥ h√†m c·∫ßn **m·ªôt** ƒë·ªëi s·ªë: `abs(-10)`
   - C√≥ h√†m c·∫ßn **nhi·ªÅu** ƒë·ªëi s·ªë: `round(3.14159, 2)`
   - C√≥ h√†m **kh√¥ng c·∫ßn ƒë·ªëi s·ªë**: `input()` khi d√πng m·∫∑c ƒë·ªãnh

   üëâ Quan tr·ªçng:  **ƒë·ªëi s·ªë lu√¥n ƒë∆∞·ª£c truy·ªÅn v√†o h√†m th√¥ng qua d·∫•u ngo·∫∑c tr√≤n** `()` khi g·ªçi h√†m.

3. **Lu√¥n tr·∫£ v·ªÅ m·ªôt k·∫øt qu·∫£ (return value)**  
   ‚Üí Khi b·∫°n g·ªçi m·ªôt h√†m, Python s·∫Ω th·ª±c hi·ªán x·ª≠ l√Ω v√† **tr·∫£ l·∫°i m·ªôt gi√° tr·ªã n√†o ƒë√≥** ‚Äì g·ªçi l√† **gi√° tr·ªã tr·∫£ v·ªÅ** (return value).  

   K·∫øt qu·∫£ n√†y c√≥ th·ªÉ l√†:
   - M·ªôt con s·ªë ‚Üí `abs(-7)` tr·∫£ v·ªÅ **7**
   - M·ªôt chu·ªói ‚Üí `"abc".upper()` tr·∫£ v·ªÅ **"ABC"**
   - M·ªôt ki·ªÉu d·ªØ li·ªáu ‚Üí `type(5)` tr·∫£ v·ªÅ **<class 'int'>**
   - M·ªôt gi√° tr·ªã ƒë·∫∑c bi·ªát l√† `None` (kh√¥ng c√≥ gi√° tr·ªã c·ª• th·ªÉ)

   V√≠ d·ª•:
   - `len("abc")` ‚Üí tr·∫£ v·ªÅ **3** (s·ªë nguy√™n)
   - `type(5)` ‚Üí tr·∫£ v·ªÅ `<class 'int'>`
   - `print("Hello")` ‚Üí in ra m√†n h√¨nh `"Hello"` **nh∆∞ng kh√¥ng tr·∫£ v·ªÅ g√¨**, v√¨ v·∫≠y gi√° tr·ªã tr·∫£ v·ªÅ l√† `None`

   üí° `None` l√† m·ªôt ki·ªÉu d·ªØ li·ªáu ƒë·∫∑c bi·ªát trong Python, d√πng ƒë·ªÉ bi·ªÉu th·ªã **"kh√¥ng c√≥ g√¨"**. M·ªôt s·ªë h√†m kh√¥ng c·∫ßn tr·∫£ k·∫øt qu·∫£ ƒë·ªÉ x·ª≠ l√Ω ti·∫øp (nh∆∞ `print()`) th√¨ s·∫Ω m·∫∑c ƒë·ªãnh tr·∫£ v·ªÅ `None`.





```python
result = print("daiduong.eth")
print(result)  # s·∫Ω in ra: None
```

#### ‚úÖ 2. Built-in Function l√† g√¨?

Trong Python, **h√†m Built-in** l√† nh·ªØng h√†m ƒë√£ ƒë∆∞·ª£c t√≠ch h·ª£p s·∫µn ‚Äì b·∫°n c√≥ th·ªÉ **s·ª≠ d·ª•ng tr·ª±c ti·∫øp** m√† **kh√¥ng c·∫ßn t·ª± ƒë·ªãnh nghƒ©a ho·∫∑c import th∆∞ vi·ªán** n√†o.

### üìå ƒê·∫∑c ƒëi·ªÉm c·ªßa h√†m Built-in:
- ƒê∆∞·ª£c c√†i s·∫µn b·ªüi Python
- C√≥ th·ªÉ d√πng ·ªü m·ªçi n∆°i trong ch∆∞∆°ng tr√¨nh
- H·ªó tr·ª£ nhi·ªÅu thao t√°c c∆° b·∫£n: hi·ªÉn th·ªã, ki·ªÉm tra, chuy·ªÉn ƒë·ªïi, t√≠nh to√°n‚Ä¶

### üß† V√≠ d·ª•:
- `print("Hello")` ‚Üí In chu·ªói "Hello" ra m√†n h√¨nh
- `type(123)` ‚Üí Tr·∫£ v·ªÅ ki·ªÉu d·ªØ li·ªáu c·ªßa s·ªë 123
- `abs(-5)` ‚Üí Tr·∫£ v·ªÅ gi√° tr·ªã tuy·ªát ƒë·ªëi c·ªßa -5 l√† 5
- `len("data")` ‚Üí ƒê·∫øm s·ªë k√Ω t·ª± trong chu·ªói

‚õ≥Ô∏è **Ghi nh·ªõ**: B·∫°n ƒë√£ s·ª≠ d·ª•ng h√†m **Built-in** t·ª´ B√†i 1 m√† c√≥ th·ªÉ ch∆∞a ƒë·ªÉ √Ω ‚Äì v√≠ d·ª• nh∆∞ `print()` hay `type()` ch√≠nh l√† c√°c **h√†m Built-in** r·∫•t quen thu·ªôc.

| H√†m | M√¥ t·∫£ ng·∫Øn g·ªçn | V√≠ d·ª• |
|-----|----------------|-------|
| `print()` | In ra m√†n h√¨nh | `print("Hello")` |
| `type()` | Ki·ªÉm tra ki·ªÉu d·ªØ li·ªáu | `type(3.14)` ‚Üí `<class 'float'>` |
| `len()` | ƒê·∫øm ƒë·ªô d√†i chu·ªói, list, tuple‚Ä¶ | `len("data")` ‚Üí `4` |
| `int()` | √âp ki·ªÉu v·ªÅ s·ªë nguy√™n | `int("10")` ‚Üí `10` |
| `float()` | √âp ki·ªÉu v·ªÅ s·ªë th·ª±c | `float("3.5")` ‚Üí `3.5` |
| `str()` | √âp ki·ªÉu v·ªÅ chu·ªói | `str(100)` ‚Üí `"100"` |
| `bool()` | √âp ki·ªÉu v·ªÅ gi√° tr·ªã logic | `bool(0)` ‚Üí `False` |
| `round()` | L√†m tr√≤n s·ªë th·ª±c | `round(3.14159, 2)` ‚Üí `3.14` |
| `abs()` | Tr·∫£ gi√° tr·ªã tuy·ªát ƒë·ªëi | `abs(-9)` ‚Üí `9` |
| `sum()` | T√≠nh t·ªïng c√°c ph·∫ßn t·ª≠ trong list | `sum([1, 2, 3])` ‚Üí `6` |


```python
# üß™ C√°c h√†m x·ª≠ l√Ω s·ªë: abs(), round(), type()

x = -22.10  # S·ªë th·ª±c √¢m

print("Gi√° tr·ªã ban ƒë·∫ßu c·ªßa x:", x)
print("abs(x):", abs(x))          # Tr·∫£ v·ªÅ gi√° tr·ªã tuy·ªát ƒë·ªëi ‚Üí 22.1
print("round(x, 1):", round(x, 1))  # L√†m tr√≤n ƒë·∫øn 1 ch·ªØ s·ªë th·∫≠p ph√¢n ‚Üí -22.1
print("type(x):", type(x))        # Ki·ªÉu d·ªØ li·ªáu c·ªßa x l√† float
```

```python
# üìè H√†m ƒëo ƒë·ªô d√†i chu·ªói: len()

z = "MeoMeo"  # M·ªôt chu·ªói 7 k√Ω t·ª±

print("Chu·ªói z:", z)
print("len(z):", len(z))  # Tr·∫£ v·ªÅ ƒë·ªô d√†i chu·ªói ‚Üí 7
```

```python
# üîÅ √âp ki·ªÉu t·ª´ chu·ªói sang s·ªë: int(), float()

y = "2025"  # Chu·ªói th·ªÉ hi·ªán m·ªôt s·ªë

print("Chu·ªói y:", y)
print("int(y):", int(y))      # √âp th√†nh s·ªë nguy√™n ‚Üí 2025
print("float(y):", float(y))  # √âp th√†nh s·ªë th·ª±c ‚Üí 2025.0
```

```python
# üßµ √âp ki·ªÉu s·ªë v·ªÅ chu·ªói: str()

x = -22.1

print("str(x):", str(x))  # √âp s·ªë -22.1 th√†nh chu·ªói "-22.1"
```

```python
# ‚úÖ H√†m ki·ªÉm tra gi√° tr·ªã logic: bool()

print("bool(0):", bool(0))          # 0 ‚Üí False
print("bool(''):", bool(""))        # Chu·ªói r·ªóng ‚Üí False
print("bool('Xom'):", bool("Xom"))  # Chu·ªói kh√¥ng r·ªóng ‚Üí True
```

#### ‚úÖ 3. Method l√† g√¨?

Trong Python, **Method** l√† m·ªôt d·∫°ng **H√†m g·∫Øn li·ªÅn v·ªõi m·ªôt ƒë·ªëi t∆∞·ª£ng**.

N√≥i c√°ch kh√°c:
- N·∫øu `built-in function` gi·ªëng nh∆∞ **m·ªôt c√¥ng c·ª• ƒë·ªôc l·∫≠p** (v√≠ d·ª•: `len("abc")`)
- Th√¨ `method` gi·ªëng nh∆∞ **m·ªôt h√†nh ƒë·ªông c·ªßa ch√≠nh ƒë·ªëi t∆∞·ª£ng ƒë√≥** (v√≠ d·ª•: `"abc".upper()`)

###üìå C·∫•u tr√∫c g·ªçi method:
> <ƒë·ªëi_t∆∞·ª£ng>.<t√™n_ph∆∞∆°ng_th·ª©c>()


### üìå ƒê·∫∑c ƒëi·ªÉm c·ªßa method:
- Lu√¥n ƒëi k√®m theo ƒë·ªëi t∆∞·ª£ng c·ª• th·ªÉ:
‚Üí B·∫°n kh√¥ng th·ªÉ g·ªçi **upper()** m·ªôt m√¨nh ‚Äì ph·∫£i vi·∫øt **"abc".upper()**
- Ch·ªâ t·ªìn t·∫°i ·ªü nh·ªØng ki·ªÉu d·ªØ li·ªáu ph√π h·ª£p:
‚Üí V√≠ d·ª•: **str** c√≥ **.upper()**, **.replace()**, c√≤n **int** th√¨ kh√¥ng.
- C√≥ th·ªÉ nh·∫≠n ƒë·ªëi s·ªë **(argument)** v√† tr·∫£ v·ªÅ k·∫øt qu·∫£ **(return value)**:
‚Üí Gi·ªëng nh∆∞ h√†m th√¥ng th∆∞·ªùng, nhi·ªÅu **method** cho ph√©p b·∫°n truy·ªÅn tham s·ªë.



```python
s = "hello".upper()  # G·ªçi method upper() c·ªßa chu·ªói
print(s)
```

**üßµ Method cho ki·ªÉu string (chu·ªói)**

Trong Python, `str` l√† m·ªôt trong nh·ªØng ki·ªÉu d·ªØ li·ªáu quan tr·ªçng v√† ph·ªï bi·∫øn nh·∫•t ‚Äì ƒë·∫∑c bi·ªát khi x·ª≠ l√Ω d·ªØ li·ªáu vƒÉn b·∫£n (t√™n, ƒë·ªãa ch·ªâ, m√¥ t·∫£‚Ä¶).

Python cung c·∫•p **r·∫•t nhi·ªÅu method h·ªØu √≠ch** cho string.

üìå M·ªôt s·ªë method b·∫°n n√™n bi·∫øt:

| Method | Ch·ª©c nƒÉng |
|--------|-----------|
| `.upper()` / `.lower()` | Vi·∫øt hoa / vi·∫øt th∆∞·ªùng to√†n b·ªô |
| `.replace(old, new)` | Thay th·∫ø chu·ªói con |
| `.strip()` | X√≥a kho·∫£ng tr·∫Øng ƒë·∫ßu/cu·ªëi chu·ªói |
| `.startswith()` / `.endswith()` | Ki·ªÉm tra chu·ªói c√≥ b·∫Øt ƒë·∫ßu/k·∫øt th√∫c b·∫±ng k√Ω t·ª± n√†o ƒë√≥ |
| `.isalpha()` / `.isdigit()` | Ki·ªÉm tra chu·ªói c√≥ ch·ªâ g·ªìm ch·ªØ / s·ªë |
| `.find(sub)` / `.index(sub)` | T√¨m v·ªã tr√≠ xu·∫•t hi·ªán chu·ªói con |
| `.count(sub)` | ƒê·∫øm s·ªë l·∫ßn xu·∫•t hi·ªán chu·ªói con |


```python
s = "  XomData2025  "

print("Chu·ªói g·ªëc:", s)

# Vi·∫øt hoa v√† vi·∫øt th∆∞·ªùng
print("s.upper():", s.upper())
print("s.lower():", s.lower())

# X√≥a kho·∫£ng tr·∫Øng ƒë·∫ßu-cu·ªëi
print("s.strip():", s.strip())

# Thay th·∫ø chu·ªói
print("s.replace('2025', 'Python'):", s.replace("2025", "Python"))

# Ki·ªÉm tra b·∫Øt ƒë·∫ßu/k·∫øt th√∫c
print("s.startswith('Xom'):", s.startswith("Xom"))         # False (v√¨ c√≥ kho·∫£ng tr·∫Øng ƒë·∫ßu)
print("s.strip().startswith('Xom'):", s.strip().startswith("Xom"))  # True
```

```python
s = "PythonData123"

# Ki·ªÉm tra chu·ªói
print("s.isalpha():", s.isalpha())   # False (v√¨ c√≥ s·ªë)
print("s.isdigit():", s.isdigit())   # False

# T√¨m ki·∫øm chu·ªói con
print("s.find('Data'):", s.find("Data"))   # Tr·∫£ v·ªÅ v·ªã tr√≠ b·∫Øt ƒë·∫ßu
print("s.count('a'):", s.count("a"))       # ƒê·∫øm s·ªë l·∫ßn xu·∫•t hi·ªán
```

üéÅ **BONUS: M·ªôt s·ªë method v√† x·ª≠ l√Ω m·ªü r·ªông v·ªõi `int`, `float`, `bool`**

Trong ph√¢n t√≠ch d·ªØ li·ªáu, ki·ªÉu `int`, `float`, `bool` **√≠t c√≥ method th·ª±c t·∫ø c·∫ßn d√πng**, nh∆∞ng n·∫øu b·∫°n t√≤ m√≤ th√¨ d∆∞·ªõi ƒë√¢y l√† m·ªôt s·ªë v√≠ d·ª• ƒë∆°n gi·∫£n c√≥ th·ªÉ th·ª≠.

```python
# üéØ M·ªôt s·ªë method v√† x·ª≠ l√Ω m·ªü r·ªông v·ªõi ki·ªÉu int

a = 15  # s·ªë nguy√™n

# Method 1: conjugate()
# Tr·∫£ v·ªÅ s·ªë li√™n h·ª£p (d√†nh cho s·ªë ph·ª©c). V·ªõi int th√¨ tr·∫£ v·ªÅ ch√≠nh n√≥.
print("a.conjugate():", a.conjugate())  # K·∫øt qu·∫£: 15

# Method 2: to_bytes()
# Chuy·ªÉn s·ªë nguy√™n th√†nh d·∫°ng byte. Th∆∞·ªùng d√πng trong x·ª≠ l√Ω nh·ªã ph√¢n.
# C√∫ ph√°p: s·ªë.to_bytes(s·ªë_byte, endian)
print("a.to_bytes(2, 'big'):", a.to_bytes(2, 'big'))  # K·∫øt qu·∫£: b'\x00\x0f'

# Method 3: bit_length()
# Tr·∫£ v·ªÅ s·ªë bit c·∫ßn thi·∫øt ƒë·ªÉ bi·ªÉu di·ªÖn s·ªë nguy√™n n√†y d∆∞·ªõi d·∫°ng nh·ªã ph√¢n (binary)
# 15 -> nh·ªã ph√¢n l√† 1111 => c·∫ßn 4 bit
print("a.bit_length():", a.bit_length())  # K·∫øt qu·∫£: 4
```

```python
# üéØ M·ªôt s·ªë method v√† x·ª≠ l√Ω m·ªü r·ªông v·ªõi ki·ªÉu float

b = 3.14  # s·ªë th·ª±c
c = 7.0   # s·ªë th·ª±c nh∆∞ng c√≥ ph·∫ßn th·∫≠p ph√¢n l√† .0

# Method 1: is_integer()
# Ki·ªÉm tra xem s·ªë th·ª±c n√†y c√≥ ph·∫£i l√† m·ªôt s·ªë nguy√™n kh√¥ng (v√≠ d·ª• 7.0 l√† nguy√™n)
print("c.is_integer():", c.is_integer())  # K·∫øt qu·∫£: True
print("b.is_integer():", b.is_integer())  # K·∫øt qu·∫£: False

# Method 2: hex()
# Tr·∫£ v·ªÅ bi·ªÉu di·ªÖn s·ªë th·ª±c n√†y d∆∞·ªõi d·∫°ng h·ªá th·∫≠p l·ª•c ph√¢n (hexadecimal)
print("b.hex():", b.hex())  # V√≠ d·ª•: '0x1.91eb851eb851fp+1'

# Method 3: as_integer_ratio()
# Tr·∫£ v·ªÅ ph√¢n s·ªë bi·ªÉu di·ªÖn ch√≠nh x√°c s·ªë th·ª±c n√†y d∆∞·ªõi d·∫°ng (t·ª≠, m·∫´u)
print("b.as_integer_ratio():", b.as_integer_ratio())  # K·∫øt qu·∫£: (3537115888337719, 1125899906842624)
```

```python
# üéØ M·ªôt s·ªë method v√† thao t√°c m·ªü r·ªông v·ªõi ki·ªÉu bool

a = True
b = False

# Method 1: bit_length()
# Do bool k·∫ø th·ª´a t·ª´ int n√™n c≈©ng c√≥ method bit_length()
# True = 1 => bit_length = 1; False = 0 => bit_length = 0
print("a.bit_length():", a.bit_length())  # K·∫øt qu·∫£: 1
print("b.bit_length():", b.bit_length())  # K·∫øt qu·∫£: 0

# Method 2: conjugate()
# C≈©ng k·∫ø th·ª´a t·ª´ int
print("a.conjugate():", a.conjugate())  # K·∫øt qu·∫£: True
print("b.conjugate():", b.conjugate())  # K·∫øt qu·∫£: False
```
