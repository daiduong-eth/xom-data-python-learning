## **📚 F_01. Python Function (Hàm trong Python)**

Trong Python, **hàm (function)** là một công cụ giúp bạn **thực hiện một công việc cụ thể**: xử lý dữ liệu, tính toán, chuyển đổi kiểu, kiểm tra điều kiện...
sau đó **trả về một giá trị nào đó**.

#### ✅ 1. Đặc điểm của hàm

1. **Thực hiện một hành động cụ thể**  
   → Ví dụ: `print()` để in ra màn hình, `type()` để kiểm tra kiểu, `abs()` để lấy giá trị tuyệt đối của các giá trị số.

2. **Có thể nhận đầu vào (gọi là đối số – arguments)**  
   → Khi bạn sử dụng một hàm, thường bạn sẽ **truyền dữ liệu vào trong dấu ngoặc** của hàm đó. Dữ liệu được truyền vào này được gọi là **đối số** (argument).

   Ví dụ: `len("Hello")`  
   - `len` là tên của **hàm** – có nhiệm vụ đo độ dài của một đối tượng (chuỗi, danh sách, v.v.)
   - `"Hello"` là một **giá trị có kiểu (string)** được truyền vào hàm – chính là **đối số**  
   → Khi chạy `len("Hello")`, Python sẽ trả về kết quả **5**, vì chuỗi `"Hello"` có 5 ký tự.

   🧠 Không phải hàm nào cũng cần đối số:
   - Có hàm cần **một** đối số: `abs(-10)`
   - Có hàm cần **nhiều** đối số: `round(3.14159, 2)`
   - Có hàm **không cần đối số**: `input()` khi dùng mặc định

   👉 Quan trọng:  **đối số luôn được truyền vào hàm thông qua dấu ngoặc tròn** `()` khi gọi hàm.

3. **Luôn trả về một kết quả (return value)**  
   → Khi bạn gọi một hàm, Python sẽ thực hiện xử lý và **trả lại một giá trị nào đó** – gọi là **giá trị trả về** (return value).  

   Kết quả này có thể là:
   - Một con số → `abs(-7)` trả về **7**
   - Một chuỗi → `"abc".upper()` trả về **"ABC"**
   - Một kiểu dữ liệu → `type(5)` trả về **<class 'int'>**
   - Một giá trị đặc biệt là `None` (không có giá trị cụ thể)

   Ví dụ:
   - `len("abc")` → trả về **3** (số nguyên)
   - `type(5)` → trả về `<class 'int'>`
   - `print("Hello")` → in ra màn hình `"Hello"` **nhưng không trả về gì**, vì vậy giá trị trả về là `None`

   💡 `None` là một kiểu dữ liệu đặc biệt trong Python, dùng để biểu thị **"không có gì"**. Một số hàm không cần trả kết quả để xử lý tiếp (như `print()`) thì sẽ mặc định trả về `None`.





```python
result = print("daiduong.eth")
print(result)  # sẽ in ra: None
```

#### ✅ 2. Built-in Function là gì?

Trong Python, **hàm Built-in** là những hàm đã được tích hợp sẵn – bạn có thể **sử dụng trực tiếp** mà **không cần tự định nghĩa hoặc import thư viện** nào.

### 📌 Đặc điểm của hàm Built-in:
- Được cài sẵn bởi Python
- Có thể dùng ở mọi nơi trong chương trình
- Hỗ trợ nhiều thao tác cơ bản: hiển thị, kiểm tra, chuyển đổi, tính toán…

### 🧠 Ví dụ:
- `print("Hello")` → In chuỗi "Hello" ra màn hình
- `type(123)` → Trả về kiểu dữ liệu của số 123
- `abs(-5)` → Trả về giá trị tuyệt đối của -5 là 5
- `len("data")` → Đếm số ký tự trong chuỗi

⛳️ **Ghi nhớ**: Bạn đã sử dụng hàm **Built-in** từ Bài 1 mà có thể chưa để ý – ví dụ như `print()` hay `type()` chính là các **hàm Built-in** rất quen thuộc.

| Hàm | Mô tả ngắn gọn | Ví dụ |
|-----|----------------|-------|
| `print()` | In ra màn hình | `print("Hello")` |
| `type()` | Kiểm tra kiểu dữ liệu | `type(3.14)` → `<class 'float'>` |
| `len()` | Đếm độ dài chuỗi, list, tuple… | `len("data")` → `4` |
| `int()` | Ép kiểu về số nguyên | `int("10")` → `10` |
| `float()` | Ép kiểu về số thực | `float("3.5")` → `3.5` |
| `str()` | Ép kiểu về chuỗi | `str(100)` → `"100"` |
| `bool()` | Ép kiểu về giá trị logic | `bool(0)` → `False` |
| `round()` | Làm tròn số thực | `round(3.14159, 2)` → `3.14` |
| `abs()` | Trả giá trị tuyệt đối | `abs(-9)` → `9` |
| `sum()` | Tính tổng các phần tử trong list | `sum([1, 2, 3])` → `6` |


```python
# 🧪 Các hàm xử lý số: abs(), round(), type()

x = -22.10  # Số thực âm

print("Giá trị ban đầu của x:", x)
print("abs(x):", abs(x))          # Trả về giá trị tuyệt đối → 22.1
print("round(x, 1):", round(x, 1))  # Làm tròn đến 1 chữ số thập phân → -22.1
print("type(x):", type(x))        # Kiểu dữ liệu của x là float
```

```python
# 📏 Hàm đo độ dài chuỗi: len()

z = "MeoMeo"  # Một chuỗi 7 ký tự

print("Chuỗi z:", z)
print("len(z):", len(z))  # Trả về độ dài chuỗi → 7
```

```python
# 🔁 Ép kiểu từ chuỗi sang số: int(), float()

y = "2025"  # Chuỗi thể hiện một số

print("Chuỗi y:", y)
print("int(y):", int(y))      # Ép thành số nguyên → 2025
print("float(y):", float(y))  # Ép thành số thực → 2025.0
```

```python
# 🧵 Ép kiểu số về chuỗi: str()

x = -22.1

print("str(x):", str(x))  # Ép số -22.1 thành chuỗi "-22.1"
```

```python
# ✅ Hàm kiểm tra giá trị logic: bool()

print("bool(0):", bool(0))          # 0 → False
print("bool(''):", bool(""))        # Chuỗi rỗng → False
print("bool('Xom'):", bool("Xom"))  # Chuỗi không rỗng → True
```

#### ✅ 3. Method là gì?

Trong Python, **Method** là một dạng **Hàm gắn liền với một đối tượng**.

Nói cách khác:
- Nếu `built-in function` giống như **một công cụ độc lập** (ví dụ: `len("abc")`)
- Thì `method` giống như **một hành động của chính đối tượng đó** (ví dụ: `"abc".upper()`)

###📌 Cấu trúc gọi method:
> <đối_tượng>.<tên_phương_thức>()


### 📌 Đặc điểm của method:
- Luôn đi kèm theo đối tượng cụ thể:
→ Bạn không thể gọi **upper()** một mình – phải viết **"abc".upper()**
- Chỉ tồn tại ở những kiểu dữ liệu phù hợp:
→ Ví dụ: **str** có **.upper()**, **.replace()**, còn **int** thì không.
- Có thể nhận đối số **(argument)** và trả về kết quả **(return value)**:
→ Giống như hàm thông thường, nhiều **method** cho phép bạn truyền tham số.



```python
s = "hello".upper()  # Gọi method upper() của chuỗi
print(s)
```

**🧵 Method cho kiểu string (chuỗi)**

Trong Python, `str` là một trong những kiểu dữ liệu quan trọng và phổ biến nhất – đặc biệt khi xử lý dữ liệu văn bản (tên, địa chỉ, mô tả…).

Python cung cấp **rất nhiều method hữu ích** cho string.

📌 Một số method bạn nên biết:

| Method | Chức năng |
|--------|-----------|
| `.upper()` / `.lower()` | Viết hoa / viết thường toàn bộ |
| `.replace(old, new)` | Thay thế chuỗi con |
| `.strip()` | Xóa khoảng trắng đầu/cuối chuỗi |
| `.startswith()` / `.endswith()` | Kiểm tra chuỗi có bắt đầu/kết thúc bằng ký tự nào đó |
| `.isalpha()` / `.isdigit()` | Kiểm tra chuỗi có chỉ gồm chữ / số |
| `.find(sub)` / `.index(sub)` | Tìm vị trí xuất hiện chuỗi con |
| `.count(sub)` | Đếm số lần xuất hiện chuỗi con |


```python
s = "  XomData2025  "

print("Chuỗi gốc:", s)

# Viết hoa và viết thường
print("s.upper():", s.upper())
print("s.lower():", s.lower())

# Xóa khoảng trắng đầu-cuối
print("s.strip():", s.strip())

# Thay thế chuỗi
print("s.replace('2025', 'Python'):", s.replace("2025", "Python"))

# Kiểm tra bắt đầu/kết thúc
print("s.startswith('Xom'):", s.startswith("Xom"))         # False (vì có khoảng trắng đầu)
print("s.strip().startswith('Xom'):", s.strip().startswith("Xom"))  # True
```

```python
s = "PythonData123"

# Kiểm tra chuỗi
print("s.isalpha():", s.isalpha())   # False (vì có số)
print("s.isdigit():", s.isdigit())   # False

# Tìm kiếm chuỗi con
print("s.find('Data'):", s.find("Data"))   # Trả về vị trí bắt đầu
print("s.count('a'):", s.count("a"))       # Đếm số lần xuất hiện
```

🎁 **BONUS: Một số method và xử lý mở rộng với `int`, `float`, `bool`**

Trong phân tích dữ liệu, kiểu `int`, `float`, `bool` **ít có method thực tế cần dùng**, nhưng nếu bạn tò mò thì dưới đây là một số ví dụ đơn giản có thể thử.

```python
# 🎯 Một số method và xử lý mở rộng với kiểu int

a = 15  # số nguyên

# Method 1: conjugate()
# Trả về số liên hợp (dành cho số phức). Với int thì trả về chính nó.
print("a.conjugate():", a.conjugate())  # Kết quả: 15

# Method 2: to_bytes()
# Chuyển số nguyên thành dạng byte. Thường dùng trong xử lý nhị phân.
# Cú pháp: số.to_bytes(số_byte, endian)
print("a.to_bytes(2, 'big'):", a.to_bytes(2, 'big'))  # Kết quả: b'\x00\x0f'

# Method 3: bit_length()
# Trả về số bit cần thiết để biểu diễn số nguyên này dưới dạng nhị phân (binary)
# 15 -> nhị phân là 1111 => cần 4 bit
print("a.bit_length():", a.bit_length())  # Kết quả: 4
```

```python
# 🎯 Một số method và xử lý mở rộng với kiểu float

b = 3.14  # số thực
c = 7.0   # số thực nhưng có phần thập phân là .0

# Method 1: is_integer()
# Kiểm tra xem số thực này có phải là một số nguyên không (ví dụ 7.0 là nguyên)
print("c.is_integer():", c.is_integer())  # Kết quả: True
print("b.is_integer():", b.is_integer())  # Kết quả: False

# Method 2: hex()
# Trả về biểu diễn số thực này dưới dạng hệ thập lục phân (hexadecimal)
print("b.hex():", b.hex())  # Ví dụ: '0x1.91eb851eb851fp+1'

# Method 3: as_integer_ratio()
# Trả về phân số biểu diễn chính xác số thực này dưới dạng (tử, mẫu)
print("b.as_integer_ratio():", b.as_integer_ratio())  # Kết quả: (3537115888337719, 1125899906842624)
```

```python
# 🎯 Một số method và thao tác mở rộng với kiểu bool

a = True
b = False

# Method 1: bit_length()
# Do bool kế thừa từ int nên cũng có method bit_length()
# True = 1 => bit_length = 1; False = 0 => bit_length = 0
print("a.bit_length():", a.bit_length())  # Kết quả: 1
print("b.bit_length():", b.bit_length())  # Kết quả: 0

# Method 2: conjugate()
# Cũng kế thừa từ int
print("a.conjugate():", a.conjugate())  # Kết quả: True
print("b.conjugate():", b.conjugate())  # Kết quả: False
```
