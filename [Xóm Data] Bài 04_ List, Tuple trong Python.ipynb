#### üîç 3. Truy c·∫≠p ph·∫ßn t·ª≠ trong List: Indexing v√† Slicing

üìå **3.1 Indexing** ‚Äì Truy c·∫≠p ph·∫ßn t·ª≠ qua ch·ªâ s·ªë
- Khi b·∫°n mu·ªën **truy c·∫≠p m·ªôt ph·∫ßn t·ª≠ c·ª• th·ªÉ trong list**, b·∫°n d√πng c√∫ ph√°p `[]` (d·∫•u ngo·∫∑c vu√¥ng).
- Ch·ªâ s·ªë trong Python b·∫Øt ƒë·∫ßu t·ª´ **0**
- C√≥ th·ªÉ d√πng **ch·ªâ s·ªë √¢m** ƒë·ªÉ ƒë·∫øm ng∆∞·ª£c t·ª´ cu·ªëi danh s√°ch


üìç**C√∫ ph√°p c∆° b·∫£n**:
> my_list[index]

trong ƒë√≥ `index` l√† v·ªã tr√≠ ph·∫ßn t·ª≠ b·∫°n mu·ªën truy c·∫≠p

````python
my_list = ['a', 'b', 'c', 'd', 'e']

# Truy c·∫≠p b·∫±ng ch·ªâ s·ªë d∆∞∆°ng
print("Ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n:", my_list[0])   # 'a'
print("Ph·∫ßn t·ª≠ th·ª© ba:", my_list[2])     # 'c'

# Truy c·∫≠p b·∫±ng ch·ªâ s·ªë √¢m
print("Ph·∫ßn t·ª≠ cu·ªëi c√πng:", my_list[-1]) # 'e'
print("Ph·∫ßn t·ª≠ √°p ch√≥t:", my_list[-2])   # 'd'
print("Ph·∫ßn t·ª≠ th·ª© ba t·ª´ cu·ªëi:", my_list[-3])  # 'c'

# L·ªói n·∫øu index v∆∞·ª£t ph·∫°m vi
try:
    print(my_list[10])
except IndexError as e:
    print("L·ªói:", e)
````

‚úÇÔ∏è **3.2 Slicing** ‚Äì C·∫Øt ra m·ªôt ph·∫ßn c·ªßa List

**Slicing** l√† c√°ch tr√≠ch xu·∫•t **m·ªôt ƒëo·∫°n con (sub-list)** t·ª´ list g·ªëc, m√† **kh√¥ng l√†m thay ƒë·ªïi list ban ƒë·∫ßu**.


üìç**C√∫ ph√°p c∆° b·∫£n**:
> my_list[start:stop:step]

*   start: ch·ªâ s·ªë b·∫Øt ƒë·∫ßu (bao g·ªìm)
*   stop: ch·ªâ s·ªë k·∫øt th√∫c (kh√¥ng bao g·ªìm)
*   step: b∆∞·ªõc nh·∫£y gi·ªØa c√°c ph·∫ßn t·ª≠ (m·∫∑c ƒë·ªãnh l√† 1)

**N·∫øu start ho·∫∑c stop b·ªã b·ªè tr·ªëng, Python s·∫Ω t·ª± hi·ªÉu:**

*   N·∫øu b·ªè start ‚Üí b·∫Øt ƒë·∫ßu t·ª´ ƒë·∫ßu list
*   N·∫øu b·ªè stop ‚Üí c·∫Øt t·ªõi cu·ªëi list
*   N·∫øu b·ªè c·∫£ hai ‚Üí copy to√†n b·ªô list

**M·ªôt s·ªë ƒëi·ªÉm ƒë·∫∑c bi·ªát:**
*   Kh√¥ng l√†m thay ƒë·ªïi danh s√°ch g·ªëc, slicing tr·∫£ v·ªÅ b·∫£n sao m·ªõi.
*   C√≥ th·ªÉ d√πng ch·ªâ s·ªë √¢m trong slicing ƒë·ªÉ c·∫Øt t·ª´ cu·ªëi danh s√°ch.
*   step √¢m cho ph√©p ƒë·∫£o ng∆∞·ª£c danh s√°ch.

````python
my_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

# ‚úÖ D√πng ƒë·∫ßy ƒë·ªß start, stop, step
print("T·ª´ ch·ªâ s·ªë 1 ƒë·∫øn 4 (kh√¥ng bao g·ªìm 4):", my_list[1:4])        # ['b', 'c', 'd']
print("T·ª´ ch·ªâ s·ªë 0 ƒë·∫øn 6, b∆∞·ªõc nh·∫£y 2:", my_list[0:7:2])           # ['a', 'c', 'e', 'g']

# ‚úÖ B·ªè start
print("T·ª´ ƒë·∫ßu ƒë·∫øn ch·ªâ s·ªë 3:", my_list[:4])                         # ['a', 'b', 'c', 'd']

# ‚úÖ B·ªè stop
print("T·ª´ ch·ªâ s·ªë 3 ƒë·∫øn h·∫øt:", my_list[3:])                         # ['d', 'e', 'f', 'g']

# ‚úÖ B·ªè c·∫£ start v√† stop ‚Üí copy to√†n b·ªô
print("To√†n b·ªô danh s√°ch (b·∫£n sao):", my_list[:])                 # ['a', 'b', 'c', 'd', 'e', 'f', 'g']

# ‚úÖ B∆∞·ªõc nh·∫£y kh√°c 1
print("C√°ch 2 ph·∫ßn t·ª≠:", my_list[::2])                             # ['a', 'c', 'e', 'g']
print("C√°ch 3 ph·∫ßn t·ª≠:", my_list[::3])                             # ['a', 'd', 'g']

# ‚úÖ D√πng ch·ªâ s·ªë √¢m trong slicing
print("T·ª´ ph·∫ßn t·ª≠ √°p ch√≥t t·ªõi ƒë·∫ßu (step -2):", my_list[-2::-2])   # ['f', 'd', 'b']

# ‚úÖ ƒê·∫£o ng∆∞·ª£c danh s√°ch
print("ƒê·∫£o ng∆∞·ª£c:", my_list[::-1])                                # ['g', 'f', 'e', 'd', 'c', 'b', 'a']

# ‚úÖ C·∫Øt ng∆∞·ª£c t·ª´ gi·ªØa danh s√°ch
print("T·ª´ ch·ªâ s·ªë 5 v·ªÅ 1 (kh√¥ng bao g·ªìm 1):", my_list[5:1:-1])      # ['f', 'e', 'd', 'c']
````

**üîÅ 4.7 `.reverse()` vs `reversed()` ‚Äì ƒê·∫£o ng∆∞·ª£c danh s√°ch:**

Python c√≥ **2 c√°ch ƒë·∫£o ng∆∞·ª£c list**, nh∆∞ng ho·∫°t ƒë·ªông kh√°c nhau:

**Method - `.reverse()`**
- Th·ª±c hi·ªán **tr√™n list g·ªëc**
- **L√†m thay ƒë·ªïi danh s√°ch ban ƒë·∫ßu**
- Kh√¥ng tr·∫£ v·ªÅ g√¨ (`None`)

üìå C√∫ ph√°p:
> my_list.reverse()

**Function - `reversed()`**
- Kh√¥ng thay ƒë·ªïi danh s√°ch ban ƒë·∫ßu
- Tr·∫£ v·ªÅ m·ªôt ƒë·ªëi t∆∞·ª£ng iterator, c√≥ th·ªÉ duy·ªát ho·∫∑c chuy·ªÉn th√†nh list m·ªõi
- Th∆∞·ªùng d√πng ƒë·ªÉ duy·ªát ƒë·∫£o ng∆∞·ª£c ho·∫∑c t·∫°o b·∫£n sao ƒë·∫£o ng∆∞·ª£c

üìå C√∫ ph√°p:
> reversed_list = list(reversed(my_list))




````python
nums = [1, 2, 3, 4, 5]
# S·ª≠ d·ª•ng .reverse() ‚Üí thay ƒë·ªïi list g·ªëc
nums.reverse()
print("Sau .reverse():", nums)         # [5, 4, 3, 2, 1]
# S·ª≠ d·ª•ng reversed() ‚Üí kh√¥ng thay ƒë·ªïi list g·ªëc
nums2 = [10, 20, 30, 40]
rev = reversed(nums2) # L∆∞u √Ω gi√∫p x√≥m ·ªü ƒë√¢y rev kh√¥ng ph·∫£i l√† list, n√™n c·∫ßn √©p ki·ªÉu v·ªÅ list

print("Sau list(reversed()):", list(rev))  # [40, 30, 20, 10]
print("List g·ªëc v·∫´n gi·ªØ nguy√™n:", nums2)   # [10, 20, 30, 40]
````

#### üîç 3. Truy c·∫≠p ph·∫ßn t·ª≠ trong Tuple: Indexing v√† Slicing

üìå **3.1 Indexing** ‚Äì Truy c·∫≠p ph·∫ßn t·ª≠ qua ch·ªâ s·ªë  
- C≈©ng gi·ªëng nh∆∞ `list`, b·∫°n c√≥ th·ªÉ **truy c·∫≠p ph·∫ßn t·ª≠ trong tuple** b·∫±ng c√°ch s·ª≠ d·ª•ng d·∫•u `[]`.  
- Ch·ªâ s·ªë b·∫Øt ƒë·∫ßu t·ª´ **0**  
- H·ªó tr·ª£ **ch·ªâ s·ªë √¢m** ƒë·ªÉ truy c·∫≠p t·ª´ cu·ªëi v·ªÅ ƒë·∫ßu  

üìç**C√∫ ph√°p c∆° b·∫£n**:
> my_tuple[index]

trong ƒë√≥ `index` l√† v·ªã tr√≠ ph·∫ßn t·ª≠ b·∫°n mu·ªën truy c·∫≠p


````python
my_tuple = ('a', 'b', 'c', 'd', 'e')

# Truy c·∫≠p b·∫±ng ch·ªâ s·ªë d∆∞∆°ng
print("Ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n:", my_tuple[0])    # 'a'
print("Ph·∫ßn t·ª≠ th·ª© ba:", my_tuple[2])      # 'c'

# Truy c·∫≠p b·∫±ng ch·ªâ s·ªë √¢m
print("Ph·∫ßn t·ª≠ cu·ªëi c√πng:", my_tuple[-1])  # 'e'
print("Ph·∫ßn t·ª≠ √°p ch√≥t:", my_tuple[-2])    # 'd'
print("Ph·∫ßn t·ª≠ th·ª© ba t·ª´ cu·ªëi:", my_tuple[-3])  # 'c'

# L·ªói n·∫øu index v∆∞·ª£t ph·∫°m vi
try:
    print(my_tuple[10])
except IndexError as e:
    print("L·ªói:", e)
````

‚úÇÔ∏è **3.2 Slicing** ‚Äì C·∫Øt ra m·ªôt ph·∫ßn c·ªßa Tuple

**Slicing** l√† c√°ch tr√≠ch xu·∫•t **m·ªôt ƒëo·∫°n con (sub-tuple)** t·ª´ tuple g·ªëc, m√† **kh√¥ng l√†m thay ƒë·ªïi tuple ban ƒë·∫ßu**.

üìç**C√∫ ph√°p c∆° b·∫£n**:
> my_tuple[start:stop:step]

*   `start`: ch·ªâ s·ªë b·∫Øt ƒë·∫ßu (bao g·ªìm)
*   `stop`: ch·ªâ s·ªë k·∫øt th√∫c (kh√¥ng bao g·ªìm)
*   `step`: b∆∞·ªõc nh·∫£y gi·ªØa c√°c ph·∫ßn t·ª≠ (m·∫∑c ƒë·ªãnh l√† 1)

**N·∫øu `start` ho·∫∑c `stop` b·ªã b·ªè tr·ªëng, Python s·∫Ω t·ª± hi·ªÉu:**

*   N·∫øu b·ªè `start` ‚Üí b·∫Øt ƒë·∫ßu t·ª´ ƒë·∫ßu tuple
*   N·∫øu b·ªè `stop` ‚Üí c·∫Øt t·ªõi cu·ªëi tuple
*   N·∫øu b·ªè c·∫£ hai ‚Üí copy to√†n b·ªô tuple

**M·ªôt s·ªë ƒëi·ªÉm ƒë·∫∑c bi·ªát:**
*   Kh√¥ng l√†m thay ƒë·ªïi tuple g·ªëc, slicing tr·∫£ v·ªÅ m·ªôt **tuple m·ªõi**.
*   C√≥ th·ªÉ d√πng **ch·ªâ s·ªë √¢m** ƒë·ªÉ c·∫Øt t·ª´ cu·ªëi v·ªÅ ƒë·∫ßu.
*   `step` √¢m cho ph√©p **ƒë·∫£o ng∆∞·ª£c tuple**.

````python
my_tuple = ('a', 'b', 'c', 'd', 'e', 'f', 'g')

# ‚úÖ D√πng ƒë·∫ßy ƒë·ªß start, stop, step
print("T·ª´ ch·ªâ s·ªë 1 ƒë·∫øn 4 (kh√¥ng bao g·ªìm 4):", my_tuple[1:4])        # ('b', 'c', 'd')
print("T·ª´ ch·ªâ s·ªë 0 ƒë·∫øn 6, b∆∞·ªõc nh·∫£y 2:", my_tuple[0:7:2])           # ('a', 'c', 'e', 'g')

# ‚úÖ B·ªè start
print("T·ª´ ƒë·∫ßu ƒë·∫øn ch·ªâ s·ªë 3:", my_tuple[:4])                         # ('a', 'b', 'c', 'd')

# ‚úÖ B·ªè stop
print("T·ª´ ch·ªâ s·ªë 3 ƒë·∫øn h·∫øt:", my_tuple[3:])                         # ('d', 'e', 'f', 'g')

# ‚úÖ B·ªè c·∫£ start v√† stop ‚Üí copy to√†n b·ªô
print("To√†n b·ªô tuple (b·∫£n sao):", my_tuple[:])                     # ('a', 'b', 'c', 'd', 'e', 'f', 'g')

# ‚úÖ B∆∞·ªõc nh·∫£y kh√°c 1
print("C√°ch 2 ph·∫ßn t·ª≠:", my_tuple[::2])                             # ('a', 'c', 'e', 'g')
print("C√°ch 3 ph·∫ßn t·ª≠:", my_tuple[::3])                             # ('a', 'd', 'g')

# ‚úÖ D√πng ch·ªâ s·ªë √¢m trong slicing
print("T·ª´ ph·∫ßn t·ª≠ √°p ch√≥t t·ªõi ƒë·∫ßu (step -2):", my_tuple[-2::-2])   # ('f', 'd', 'b')

# ‚úÖ ƒê·∫£o ng∆∞·ª£c tuple
print("ƒê·∫£o ng∆∞·ª£c:", my_tuple[::-1])                                # ('g', 'f', 'e', 'd', 'c', 'b', 'a')

# ‚úÖ C·∫Øt ng∆∞·ª£c t·ª´ gi·ªØa tuple
print("T·ª´ ch·ªâ s·ªë 5 v·ªÅ 1 (kh√¥ng bao g·ªìm 1):", my_tuple[5:1:-1])      # ('f', 'e', 'd', 'c')
````

#### üîç 4. M·ªôt s·ªë Function v√† Method hay d√πng trong Tuple

**‚úÖ 4.1. C√°c h√†m built-in c√≥ th·ªÉ d√πng v·ªõi `tuple`**

| H√†m         | M√¥ t·∫£ ng·∫Øn                                                  |
|-------------|-------------------------------------------------------------|
| `len()`     | Tr·∫£ v·ªÅ ƒë·ªô d√†i                                               |
| `max()`     | Ph·∫ßn t·ª≠ l·ªõn nh·∫•t                                            |
| `min()`     | Ph·∫ßn t·ª≠ nh·ªè nh·∫•t                                            |
| `sum()`     | T·ªïng c√°c ph·∫ßn t·ª≠ (n·∫øu l√† s·ªë)                               |
| `sorted()`  | Tr·∫£ v·ªÅ b·∫£n sao ƒë√£ s·∫Øp x·∫øp (list m·ªõi, **kh√¥ng s·ª≠a tuple**)   |
| `tuple()`   | Chuy·ªÉn ki·ªÉu d·ªØ li·ªáu th√†nh tuple                            |


````python
data = (7, 2, 9, 4)

print("ƒê·ªô d√†i:", len(data))
print("Gi√° tr·ªã l·ªõn nh·∫•t:", max(data))
print("Gi√° tr·ªã nh·ªè nh·∫•t:", min(data))
print("T·ªïng:", sum(data))
print("S·∫Øp x·∫øp:", sorted(data))        # Tr·∫£ v·ªÅ list
print("Chuy·ªÉn list th√†nh tuple:", tuple([1, 2, 3]))  # (1, 2, 3)
````

**‚úÖ 4.2. C√°c method c√≥ s·∫µn c·ªßa `tuple`**

Kh√°c v·ªõi `list`, ki·ªÉu d·ªØ li·ªáu `tuple` ch·ªâ h·ªó tr·ª£ **2 method c∆° b·∫£n**:

| Method          | M√¥ t·∫£                                                       |
|------------------|-------------------------------------------------------------|
| `.count(x)`      | ƒê·∫øm s·ªë l·∫ßn ph·∫ßn t·ª≠ `x` xu·∫•t hi·ªán trong tuple                |
| `.index(x)`      | Tr·∫£ v·ªÅ v·ªã tr√≠ ƒë·∫ßu ti√™n m√† ph·∫ßn t·ª≠ `x` xu·∫•t hi·ªán             |

üìå Do tuple l√† **immutable**, n√™n **kh√¥ng c√≥ c√°c method nh∆∞ `.append()`, `.remove()`, `.sort()`...**


````python
my_tuple = (1, 2, 2, 3)
print("S·ªë l·∫ßn xu·∫•t hi·ªán s·ªë 2:", my_tuple.count(2))     # 2
print("V·ªã tr√≠ ƒë·∫ßu ti√™n c·ªßa s·ªë 3:", my_tuple.index(3))  # 3
````

#### üîç 5. M·ªôt s·ªë To√°n t·ª≠ hay d√πng trong Tuple


| To√°n t·ª≠       | √ù nghƒ©a                                      | V√≠ d·ª•                             |
|---------------|----------------------------------------------|------------------------------------|
| `+`           | N·ªëi hai tuple l·∫°i v·ªõi nhau                   | `(1, 2) + (3, 4)` ‚Üí `(1, 2, 3, 4)` |
| `*`           | L·∫∑p l·∫°i tuple nhi·ªÅu l·∫ßn                      | `(1, 2) * 2` ‚Üí `(1, 2, 1, 2)`      |
| `in`          | Ki·ªÉm tra ph·∫ßn t·ª≠ c√≥ trong tuple kh√¥ng        | `2 in (1, 2, 3)` ‚Üí `True`          |
| `==`, `!=`    | So s√°nh n·ªôi dung hai tuple                   | `(1, 2) == (1, 2)` ‚Üí `True`        |

üìå Do tuple l√† **immutable**, n√™n kh√¥ng c√≥ c√°c to√°n t·ª≠ g√°n (`+=`, `*=`) ƒë·ªÉ thay ƒë·ªïi n·ªôi dung t·∫°i ch·ªó.


````python
# To√°n t·ª≠ +
t1 = (1, 2)
t2 = (3, 4)
print("N·ªëi tuple:", t1 + t2)              # (1, 2, 3, 4)

# To√°n t·ª≠ *
print("L·∫∑p l·∫°i tuple:", t1 * 2)           # (1, 2, 1, 2)

# To√°n t·ª≠ in
print("Ki·ªÉm tra 2 in t1:", 2 in t1)       # True
print("Ki·ªÉm tra 5 in t2:", 5 in t2)       # False

# So s√°nh == v√† !=
t3 = (1, 2)
print("t1 == t3:", t1 == t3)              # True
print("t1 != t2:", t1 != t2)              # True
````
