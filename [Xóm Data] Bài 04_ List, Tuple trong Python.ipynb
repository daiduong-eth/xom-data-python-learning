## I. LIST

#### 📦 1. List là gì?

**List** (danh sách) là một kiểu dữ liệu có cấu trúc trong Python,  
dùng để lưu trữ một tập hợp các giá trị theo thứ tự xác định.

📌 Cú pháp khai báo:
```python
my_list = [giá_trị_1, giá_trị_2, ..., giá_trị_n]


#### 📦 2. Đặc điểm nổi bật của List

Dưới đây là những đặc điểm quan trọng giúp List trở nên linh hoạt và phổ biến:

1. ✅ **Có thứ tự (Ordered)**  
2. ✅ **Cho phép phần tử trùng lặp**  
3. ✅ **Có thể chứa nhiều kiểu dữ liệu**  
4. ✅ **Có thể lồng nhau (Nested List)**  
5. ✅ **Có thể thay đổi (Mutable)**

````python
#1. ✅ Có thứ tự
my_list = ["a", "b", "c"]
print("Phần tử đầu tiên:", my_list[0])  # luôn là "a"
print("Phần tử cuối cùng:", my_list[2])  # luôn là "c"
````

````python
#2. ✅ Cho phép phần tử trùng lặp
dup_list = [1, 2, 2, 3, 3, 3]
print("List có phần tử trùng:", dup_list)
````

````python
#3. ✅ Có thể chứa nhiều kiểu dữ liệu
mixed = [123, "Python", 3.14, True]
for i, item in enumerate(mixed):
    print(f"Phần tử {i}: {item} – Kiểu: {type(item)}")
````

````python
#4. ✅ Có thể lồng nhau
nested = [[1, 2], ["a", "b"], [True, False]]
print("Phần tử thứ 2:", nested[1])          # ['a', 'b']
print("Truy cập phần tử bên trong:", nested[1][0])  # 'a'
````

````python
#5. ✅ Có thể thay đổi
nums = [10, 20, 30]
nums[1] = 99
print("Sau khi thay đổi phần tử thứ 2:", nums)
nums.append(40)
print("Sau khi thêm phần tử:", nums)
````

#### 🔍 3. Truy cập phần tử trong List: Indexing và Slicing

📌 **3.1 Indexing** – Truy cập phần tử qua chỉ số
- Khi bạn muốn **truy cập một phần tử cụ thể trong list**, bạn dùng cú pháp `[]` (dấu ngoặc vuông).
- Chỉ số trong Python bắt đầu từ **0**
- Có thể dùng **chỉ số âm** để đếm ngược từ cuối danh sách


📍**Cú pháp cơ bản**:
> my_list[index]

trong đó `index` là vị trí phần tử bạn muốn truy cập

````python
my_list = ['a', 'b', 'c', 'd', 'e']

# Truy cập bằng chỉ số dương
print("Phần tử đầu tiên:", my_list[0])   # 'a'
print("Phần tử thứ ba:", my_list[2])     # 'c'

# Truy cập bằng chỉ số âm
print("Phần tử cuối cùng:", my_list[-1]) # 'e'
print("Phần tử áp chót:", my_list[-2])   # 'd'
print("Phần tử thứ ba từ cuối:", my_list[-3])  # 'c'

# Lỗi nếu index vượt phạm vi
try:
    print(my_list[10])
except IndexError as e:
    print("Lỗi:", e)
````

✂️ **3.2 Slicing** – Cắt ra một phần của List

**Slicing** là cách trích xuất **một đoạn con (sub-list)** từ list gốc, mà **không làm thay đổi list ban đầu**.


📍**Cú pháp cơ bản**:
> my_list[start:stop:step]

*   start: chỉ số bắt đầu (bao gồm)
*   stop: chỉ số kết thúc (không bao gồm)
*   step: bước nhảy giữa các phần tử (mặc định là 1)

**Nếu start hoặc stop bị bỏ trống, Python sẽ tự hiểu:**

*   Nếu bỏ start → bắt đầu từ đầu list
*   Nếu bỏ stop → cắt tới cuối list
*   Nếu bỏ cả hai → copy toàn bộ list

**Một số điểm đặc biệt:**
*   Không làm thay đổi danh sách gốc, slicing trả về bản sao mới.
*   Có thể dùng chỉ số âm trong slicing để cắt từ cuối danh sách.
*   step âm cho phép đảo ngược danh sách.

````python
my_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

# ✅ Dùng đầy đủ start, stop, step
print("Từ chỉ số 1 đến 4 (không bao gồm 4):", my_list[1:4])        # ['b', 'c', 'd']
print("Từ chỉ số 0 đến 6, bước nhảy 2:", my_list[0:7:2])           # ['a', 'c', 'e', 'g']

# ✅ Bỏ start
print("Từ đầu đến chỉ số 3:", my_list[:4])                         # ['a', 'b', 'c', 'd']

# ✅ Bỏ stop
print("Từ chỉ số 3 đến hết:", my_list[3:])                         # ['d', 'e', 'f', 'g']

# ✅ Bỏ cả start và stop → copy toàn bộ
print("Toàn bộ danh sách (bản sao):", my_list[:])                 # ['a', 'b', 'c', 'd', 'e', 'f', 'g']

# ✅ Bước nhảy khác 1
print("Cách 2 phần tử:", my_list[::2])                             # ['a', 'c', 'e', 'g']
print("Cách 3 phần tử:", my_list[::3])                             # ['a', 'd', 'g']

# ✅ Dùng chỉ số âm trong slicing
print("Từ phần tử áp chót tới đầu (step -2):", my_list[-2::-2])   # ['f', 'd', 'b']

# ✅ Đảo ngược danh sách
print("Đảo ngược:", my_list[::-1])                                # ['g', 'f', 'e', 'd', 'c', 'b', 'a']

# ✅ Cắt ngược từ giữa danh sách
print("Từ chỉ số 5 về 1 (không bao gồm 1):", my_list[5:1:-1])      # ['f', 'e', 'd', 'c']
````

**🔁 4.7 `.reverse()` vs `reversed()` – Đảo ngược danh sách:**

Python có **2 cách đảo ngược list**, nhưng hoạt động khác nhau:

**Method - `.reverse()`**
- Thực hiện **trên list gốc**
- **Làm thay đổi danh sách ban đầu**
- Không trả về gì (`None`)

📌 Cú pháp:
> my_list.reverse()

**Function - `reversed()`**
- Không thay đổi danh sách ban đầu
- Trả về một đối tượng iterator, có thể duyệt hoặc chuyển thành list mới
- Thường dùng để duyệt đảo ngược hoặc tạo bản sao đảo ngược

📌 Cú pháp:
> reversed_list = list(reversed(my_list))




````python
nums = [1, 2, 3, 4, 5]
# Sử dụng .reverse() → thay đổi list gốc
nums.reverse()
print("Sau .reverse():", nums)         # [5, 4, 3, 2, 1]
# Sử dụng reversed() → không thay đổi list gốc
nums2 = [10, 20, 30, 40]
rev = reversed(nums2) # Lưu ý giúp xóm ở đây rev không phải là list, nên cần ép kiểu về list

print("Sau list(reversed()):", list(rev))  # [40, 30, 20, 10]
print("List gốc vẫn giữ nguyên:", nums2)   # [10, 20, 30, 40]
````

#### 🔍 5. Một số Toán tử hay dùng trong List

**➕5.1 Toán tử `+` – Nối List**

Toán tử `+` dùng để **nối hai list hoặc tuple** lại với nhau, tạo thành một đối tượng mới.

**✖️ 5.2 Toán tử `*` – Lặp lại danh sách**

Toán tử `*` được dùng để **lặp lại nội dung của list hoặc tuple nhiều lần**.

**🔎 5.3 Toán tử `in` – Kiểm tra phần tử**

Toán tử `in` dùng để **kiểm tra một phần tử có tồn tại trong list/tuple hay không**.

📌 Trả về `True` hoặc `False`.


````python
fruits = ['apple', 'banana']
print('apple' in fruits)     # True
print('orange' in fruits)    # False
````

**🟰 5.4 Toán tử `==` và `!=` – So sánh danh sách**

Bạn có thể dùng `==` để **so sánh 2 list hoặc 2 tuple** về giá trị và thứ tự phần tử.  
`!=` trả về `True` nếu chúng khác nhau.

📌 Không so sánh theo địa chỉ ô nhớ, mà theo **nội dung bên trong**.


````python
a = [1, 2, 3]
b = [1, 2, 3]
c = [3, 2, 1]

print(a == b)  # True
print(a == c)  # False
print(a != c)  # True
````

## II. TUPLE

#### 🔐 1. Tuple là gì?

**Tuple** (bộ giá trị) là một kiểu dữ liệu có cấu trúc trong Python,  
dùng để lưu trữ một tập hợp các giá trị **theo thứ tự** nhưng **không thể thay đổi** sau khi được tạo ra.


📌 Thích hợp để lưu trữ các dữ liệu cố định (ví dụ: toạ độ, kích thước, hằng số, ...)

📌 Cú pháp khai báo:
>my_tuple = (giá_trị_1, giá_trị_2, ..., giá_trị_n)

#### 🔐 2. Đặc điểm nổi bật của Tuple

Dưới đây là những đặc điểm quan trọng giúp Tuple trở nên đáng tin cậy trong các trường hợp dữ liệu cố định:

1. ✅ **Có thứ tự (Ordered)**  
2. ✅ **Cho phép phần tử trùng lặp**  
3. ✅ **Có thể chứa nhiều kiểu dữ liệu**  
4. ✅ **Có thể lồng nhau (Nested Tuple)**  
5. ❌ **Không thể thay đổi (Immutable)**

````python
#1. ✅ Có thứ tự
my_tuple = ('a', 'b', 'c')
print("Phần tử đầu tiên:", my_tuple[0])   # luôn là 'a'
print("Phần tử cuối cùng:", my_tuple[2])  # luôn là 'c'
````

````python
#2. ✅ Cho phép phần tử trùng lặp
dup_tuple = (1, 2, 2, 3, 3, 3)
print("Tuple có phần tử trùng:", dup_tuple)
````

````python
#3. ✅ Có thể chứa nhiều kiểu dữ liệu
mixed = (123, "Python", 3.14, True)

for i, item in enumerate(mixed):
    print(f"Phần tử {i}: {item} – Kiểu: {type(item)}")
````

````python
#4. ✅ Có thể lồng nhau
nested = (1, 2, ('a', 'b'), (True, False))
print("Phần tử thứ 2:", nested[2])
print("Truy cập phần tử bên trong:", nested[2][0])
````

````python
#5. ❌ Không thể thay đổi
nums = (10, 20, 30)
try:
    nums[1] = 99
except TypeError as e:
    print("Lỗi:", e)
````

#### 🔍 3. Truy cập phần tử trong Tuple: Indexing và Slicing

📌 **3.1 Indexing** – Truy cập phần tử qua chỉ số  
- Cũng giống như `list`, bạn có thể **truy cập phần tử trong tuple** bằng cách sử dụng dấu `[]`.  
- Chỉ số bắt đầu từ **0**  
- Hỗ trợ **chỉ số âm** để truy cập từ cuối về đầu  

📍**Cú pháp cơ bản**:
> my_tuple[index]

trong đó `index` là vị trí phần tử bạn muốn truy cập


````python
my_tuple = ('a', 'b', 'c', 'd', 'e')

# Truy cập bằng chỉ số dương
print("Phần tử đầu tiên:", my_tuple[0])    # 'a'
print("Phần tử thứ ba:", my_tuple[2])      # 'c'

# Truy cập bằng chỉ số âm
print("Phần tử cuối cùng:", my_tuple[-1])  # 'e'
print("Phần tử áp chót:", my_tuple[-2])    # 'd'
print("Phần tử thứ ba từ cuối:", my_tuple[-3])  # 'c'

# Lỗi nếu index vượt phạm vi
try:
    print(my_tuple[10])
except IndexError as e:
    print("Lỗi:", e)
````

✂️ **3.2 Slicing** – Cắt ra một phần của Tuple

**Slicing** là cách trích xuất **một đoạn con (sub-tuple)** từ tuple gốc, mà **không làm thay đổi tuple ban đầu**.

📍**Cú pháp cơ bản**:
> my_tuple[start:stop:step]

*   `start`: chỉ số bắt đầu (bao gồm)
*   `stop`: chỉ số kết thúc (không bao gồm)
*   `step`: bước nhảy giữa các phần tử (mặc định là 1)

**Nếu `start` hoặc `stop` bị bỏ trống, Python sẽ tự hiểu:**

*   Nếu bỏ `start` → bắt đầu từ đầu tuple
*   Nếu bỏ `stop` → cắt tới cuối tuple
*   Nếu bỏ cả hai → copy toàn bộ tuple

**Một số điểm đặc biệt:**
*   Không làm thay đổi tuple gốc, slicing trả về một **tuple mới**.
*   Có thể dùng **chỉ số âm** để cắt từ cuối về đầu.
*   `step` âm cho phép **đảo ngược tuple**.

````python
my_tuple = ('a', 'b', 'c', 'd', 'e', 'f', 'g')

# ✅ Dùng đầy đủ start, stop, step
print("Từ chỉ số 1 đến 4 (không bao gồm 4):", my_tuple[1:4])        # ('b', 'c', 'd')
print("Từ chỉ số 0 đến 6, bước nhảy 2:", my_tuple[0:7:2])           # ('a', 'c', 'e', 'g')

# ✅ Bỏ start
print("Từ đầu đến chỉ số 3:", my_tuple[:4])                         # ('a', 'b', 'c', 'd')

# ✅ Bỏ stop
print("Từ chỉ số 3 đến hết:", my_tuple[3:])                         # ('d', 'e', 'f', 'g')

# ✅ Bỏ cả start và stop → copy toàn bộ
print("Toàn bộ tuple (bản sao):", my_tuple[:])                     # ('a', 'b', 'c', 'd', 'e', 'f', 'g')

# ✅ Bước nhảy khác 1
print("Cách 2 phần tử:", my_tuple[::2])                             # ('a', 'c', 'e', 'g')
print("Cách 3 phần tử:", my_tuple[::3])                             # ('a', 'd', 'g')

# ✅ Dùng chỉ số âm trong slicing
print("Từ phần tử áp chót tới đầu (step -2):", my_tuple[-2::-2])   # ('f', 'd', 'b')

# ✅ Đảo ngược tuple
print("Đảo ngược:", my_tuple[::-1])                                # ('g', 'f', 'e', 'd', 'c', 'b', 'a')

# ✅ Cắt ngược từ giữa tuple
print("Từ chỉ số 5 về 1 (không bao gồm 1):", my_tuple[5:1:-1])      # ('f', 'e', 'd', 'c')
````

#### 🔍 4. Một số Function và Method hay dùng trong Tuple

**✅ 4.1. Các hàm built-in có thể dùng với `tuple`**

| Hàm         | Mô tả ngắn                                                  |
|-------------|-------------------------------------------------------------|
| `len()`     | Trả về độ dài                                               |
| `max()`     | Phần tử lớn nhất                                            |
| `min()`     | Phần tử nhỏ nhất                                            |
| `sum()`     | Tổng các phần tử (nếu là số)                               |
| `sorted()`  | Trả về bản sao đã sắp xếp (list mới, **không sửa tuple**)   |
| `tuple()`   | Chuyển kiểu dữ liệu thành tuple                            |


````python
data = (7, 2, 9, 4)

print("Độ dài:", len(data))
print("Giá trị lớn nhất:", max(data))
print("Giá trị nhỏ nhất:", min(data))
print("Tổng:", sum(data))
print("Sắp xếp:", sorted(data))        # Trả về list
print("Chuyển list thành tuple:", tuple([1, 2, 3]))  # (1, 2, 3)
````

**✅ 4.2. Các method có sẵn của `tuple`**

Khác với `list`, kiểu dữ liệu `tuple` chỉ hỗ trợ **2 method cơ bản**:

| Method          | Mô tả                                                       |
|------------------|-------------------------------------------------------------|
| `.count(x)`      | Đếm số lần phần tử `x` xuất hiện trong tuple                |
| `.index(x)`      | Trả về vị trí đầu tiên mà phần tử `x` xuất hiện             |

📌 Do tuple là **immutable**, nên **không có các method như `.append()`, `.remove()`, `.sort()`...**


````python
my_tuple = (1, 2, 2, 3)
print("Số lần xuất hiện số 2:", my_tuple.count(2))     # 2
print("Vị trí đầu tiên của số 3:", my_tuple.index(3))  # 3
````

#### 🔍 5. Một số Toán tử hay dùng trong Tuple


| Toán tử       | Ý nghĩa                                      | Ví dụ                             |
|---------------|----------------------------------------------|------------------------------------|
| `+`           | Nối hai tuple lại với nhau                   | `(1, 2) + (3, 4)` → `(1, 2, 3, 4)` |
| `*`           | Lặp lại tuple nhiều lần                      | `(1, 2) * 2` → `(1, 2, 1, 2)`      |
| `in`          | Kiểm tra phần tử có trong tuple không        | `2 in (1, 2, 3)` → `True`          |
| `==`, `!=`    | So sánh nội dung hai tuple                   | `(1, 2) == (1, 2)` → `True`        |

📌 Do tuple là **immutable**, nên không có các toán tử gán (`+=`, `*=`) để thay đổi nội dung tại chỗ.


````python
# Toán tử +
t1 = (1, 2)
t2 = (3, 4)
print("Nối tuple:", t1 + t2)              # (1, 2, 3, 4)

# Toán tử *
print("Lặp lại tuple:", t1 * 2)           # (1, 2, 1, 2)

# Toán tử in
print("Kiểm tra 2 in t1:", 2 in t1)       # True
print("Kiểm tra 5 in t2:", 5 in t2)       # False

# So sánh == và !=
t3 = (1, 2)
print("t1 == t3:", t1 == t3)              # True
print("t1 != t2:", t1 != t2)              # True
````

## III. TỔNG KẾT

 📊 So sánh giữa List và Tuple


| Tiêu chí                       | `List`                                 | `Tuple`                                 |
|-------------------------------|----------------------------------------|------------------------------------------|
| ✅ **Cấu trúc**               | Danh sách                              | Bộ giá trị                              |
| 🔁 **Thay đổi được (mutable)**| Có                                     | ❌ Không                                 |
| 🧩 **Cú pháp khai báo**        | `[a, b, c]`                            | `(a, b, c)`                              |
| 🧠 **Có thứ tự (Ordered)**     | Có                                     | Có                                       |
| ♻️ **Cho phép trùng lặp**      | Có                                     | Có                                       |
| 🛠 **Số lượng method**         | Nhiều (append, remove, sort, ...)      | Chỉ có `count()` và `index()`           |
| 🔒 **Ứng dụng phù hợp**         | Dữ liệu thay đổi, thao tác thường xuyên| Dữ liệu cố định, không cần thay đổi     |

📌 **Tóm lại**:
- Dùng `List` khi bạn cần **thêm / xóa / sửa** dữ liệu.
- Dùng `Tuple` khi bạn muốn **bảo vệ** dữ liệu.
